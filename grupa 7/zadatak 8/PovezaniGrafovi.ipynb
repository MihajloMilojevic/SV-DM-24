# Povezani grafovi

## Šetnja

**Definicija**
Neka je $G = (V, E, \psi)$ multigraf. Neka su $e_1, \dots, e_n \in E$ i $v_0, \dots, v_n \in V$ proizvoljne grane i čvorovi sa osobinom $\psi(e_i) = \{v_{i-1}, v_i\}$, za svako $i \in \{1, \dots, n\}$. Tada za niz $$v_0 e_1 v_1 e_2 \dots e_n v_n$$ kažemo da je $v_0 v_n$-šetnja dužine $n$ u grafu $G$ između čvorova $v_0$ i $v_n$. Za čvorove $v_0$ i $v_n$ kažemo da su *krajnji čvorovi šetnje*.  


**Specijalni slučajevi**:  
1. *staza* - ako nema ponavljanja grana tj. ako za sve $i, j \in \{1, \dots, n\}$ sa osobinom $i \neq j$ važi $e_i \neq e_j$,  
2. *put* - ako nema ponavljanja čvorova tj. ako za sve $i, j \in \{0, 1, \dots, n\}$ sa osobinom $i \neq j$ važi $v_i \neq v_j$ *(osim eventualno $v_0 = v_n$)*.

Ako je graf $G$ prost, pišemo: $v_0 v_1 \dots v_n.$

Ako su krajnji čvorovi jednaki, tj. $v_0 = v_n$, tada uvodimo dodatne pojmove:
- šetnja dužine bar jedan je **zatvorena**,
- zatvorena staza je **kružna**,
- zatvoren put je **kontura**.

**Primer** Po jedan primer za graf $K_6$.
1. *šetnja* - $a e c a f d b f$;
2. *staza* - $a e c a f d b f$;
3. *put* - $a e c f d b$;
4. *kontura* - $a e f d a$.

## Povezanost

**Definicija** Neka je $G = (V, E, \psi)$ multigraf. Kažemo da su čvorovi $u$ i $v$ **povezani** ako je:
- $u = v$ ili
- $u \neq v$ i postoji $uv$-put u $G$.


Kažemo da je $G$ **povezan** akko $|V| = 1$ ili za svako $u, v \in V$ važi da su $u$ i $v$ povezani.

Postojanje $uv$-šetnje u grafu direktno implicira da su čvorovi $u$ i $v$ povezani.

**Lema** Relacija *je povezan sa* je relacija ekvivalencije na skupu čvorova grafa.

**Dokaz** Dokazujemo osobine relacija ekvivalencije (RST).

**(R)** Refleksivnost sledi direktno iz definicije.

**(S)** Neka je $u \neq v$ i neka $uv$-put u grafu oblika $uv_0 \dots v_{n-1}v$. Tada je jedan $vu$-put u grafu oblika $v v_{n-1} \dots v_0 u$.

**(T)** Pretpostavimo da u grafu $G$ postoje $uv$-put i $vw$-put: $u u_0 \dots u_{l-1} v \quad \text{i} \quad v v_0 \dots v_{n-1} w.$  Tada je sa $u u_0 \dots u_{l-1} v v_0 \dots v_{n-1} w$ data jedna $uw$-šetnja u grafu, odakle je $u$ povezan sa $w$.

**Komponenta povezanosti** grafa $G$ je povezan podgraf grafa $G$ koji nije sadržan ni u jednom drugom povezanom podgrafu grafa $G$. Drugim rečima, komponenta povezanosti grafa $G$ je **maksimalan povezan podgraf** grafa $G$.


Graf $G$ koji nije povezan ima dve ili više komponenti povezanosti koje su međusobno disjunktne i čija je unija upravo graf $G$.

Broj *komponenti povezanosti* grafa $G$, u oznaci $\omega(G)$, jednak je broju klasa ekvivalencije u odnosu na relaciju povezanosti.

**Lema** Graf $G$ je povezan akko $\omega(G) = 1$.

**Dokaz** Graf $G$ je povezan akko za svaka dva čvora $u, v \in V$ postoji $uv$-put u $G$. To dalje važi akko svi čvorovi pripadaju istoj klasi ekvivalencije u odnosu na relaciju *je povezan sa*, što važi akko $\omega(G) = 1.$

Neka je $G = (V, E, \psi)$ sa osobinom $\omega(G) = k \geq 1$.
- Čvor $v \in V$ je **razdelni (artikulacioni)** ako je $\omega(G - v) > k$,
- Grana $e \in E$ je **razdelna (most)** ako je $\omega(G - e) > k$.

**Teorema** Neka je $n \geq 2$:

Graf sa $n$ čvorova i manje od $n - 1$ grana nije povezan.

**Dokaz**: Indukcijom po $n$:


**Baza** $n = 2$: Graf sa dva čvora i 0 grana nije povezan.

**Induktivna pretpostavka**: Pretpostavimo da nijedan graf sa $n$ čvorova i manje od $n - 1$ grana nije povezan.

**Induktivni korak**: Neka je $G$ graf sa $n + 1$ čvorova i manje od $n$ grana. Postoji čvor $v$ stepena $d_G(v) \leq 1$:

- Ako je $d_G(v) = 0$: broj komponenti povezanosti je bar dva i graf nije povezan.

- Ako je $d_G(v) = 1$: $$G_0 = G - v = (V \setminus \{v\}, E \setminus \{u, v\})$$$$|V(G_0)| = n$$$$|E(G_0)| = |E(G)| - 1 < n - 1$$$$\omega(G_0) = \omega(G)$$

Prema induktivnoj pretpostavci, $\omega(G_0) \geq 2 \Rightarrow \omega(G) \geq 2$.

**Teorema** Neka je $G$ povezan i neka je $C$ kontura u grafu $G$. Ako je $e$ grana konture, onda je $G - e$ povezan.

**Dokaz** Izaberimo proizvoljno dva čvora $u, v \in V$: Kako je $G$ povezan, postoji $uv$-put $P = uv_1 \dots v_iv_{i+1} \dots v_{n-1}v$.


Imamo dve mogućnosti:

**1.** Ako $e$ ne pripada $uv$-putu, onda je $P$ $uv$-put u $G - e$.

**2.** Ako $e$ pripada $uv$-putu, onda možemo pretpostaviti da je to grana $\{v_i, v_{i+1}\}$ i da je kontura $C$ oblika $$C = v_i v_{i+1} u_1 \dots u_l v_i.$$ To znači da u grafu $G - e$ postoji put $Q$ od $v_i$ do $v_{i+1}$: $$Q = v_i u_l u_{l-1} \dots u_1 v_{i+1}.$$ Tada je šetnja u grafu $G - e$ od $u$ do $v$: $$P_2 = uv_1 \dots v_{i-1} Q v_{i+2} \dots v_{n-1} v.$$ Ako postoji šetnja u $G - e$, onda postoji $uv$-put u $G - e$. Znači, za svaka dva čvora u grafu $G - e$ postoji put koji ih povezuje.


**Definicija:** Neka je $G$ povezan graf. Rastojanje $d(u, v)$ između različitih čvorova $u$ i $v$ je dužina najkraćeg puta od $u$ do $v$. Pored toga važi $d(u, u) = 0.$

**Osobine rastojanja:**
1. $d(u, v) \geq 0$
2. $d(u, v) = 0$ ako i samo ako $u = v$
3. $d(u, v) = d(v, u)$
4. $d(u, v) + d(v, w) \geq d(u, w)$
