## **Ojlerova formula**

Kao što smo vidili u prethodnom primjeru, planarni grafovi dijele ravan na konačan broj oblasti, koje su ili neograničene ili ograničene. Ojler je dokazao da bilo koja planarna reprezentacija grafa *uvijek dijeli ravan na isti broj oblasti*. To je pokazao pronalaženjem relacije između ***broja oblasti, broja grana i broja čvorova grafa***.

#### **Teorema 1. (Ojlerova formula)** Neka je $G = (V, E)$, $|V| ≥ 2$, **povezan planaran prost graf** i neka je $f$ broj oblasti na koje graf dijeli ravan. Tada je $f = |E| - |V| + 2.$

**Dokaz.** Posmatrajmo proizvoljnu planarnu reprezentaciju grafa $G$, i neka je broj grana $|E| = n$. Dokazaćemo dato tvrđenje sukcesivnim kreiranjem podgrafova $G_1$, $G_2$, ..., $G_n$ = $G$, gdje je $G_1$ dobijen proizvoljnim izborom jedne od grana grafa $G$ i njenih incidentnih čvorova, a svaki naredni podgraf $G_m$ se dobija dodavanjem nove proizvoljne grane koja je incidentna sa jednim čvorom iz $G_{m-1}$ i eventualno dodavanjem novog čvora koji je incidentan sa tom novom granom, koja sigurno postoji jer je graf povezan. Ako pokažemo da tvrđenje važi za svako $G_k$, $k ∈ {1,2,...,n}$, to jest $f_k = |E_k| - |V_k| + 2$, gdje su $|E_k| = |E(G_k)|$ i $|V_k| = |V(G_k)|$, time je dokaz završen. U tu svrhu dokaz ćemo izvesti indukcijom.

**Baza indukcije $k = 1$:** $$f_1 = |E_1| - |V_1| + 2 = 1 - 2 + 2 = 1$$ što je tačno jer trenutno imamo samo jednu granu i dva čvora, pa ne postoji ni jedna zatvorena oblast.

**Induktivna pretpostavka:** tvrđenje važi za svako $k$, to jest $$f_k = |E_k| - |V_k| + 2.$$

**Induktivni korak:** Treba pokazati da tvrđenje važi i za $k + 1.$ Pri dodavanju nove grane, $\{u, v\}$, postoje tačno dva slučaja i to da se oba čvora već nalaze u grafu $G_k$ ili da se nalazi tačno jedan čvor, recimo $u$, a da čvor $v$ u graf $G_k$ uvodi ova nova grana. Neka je $G_{k+1} = G_k + \{u, v\}.$

**$I)$** Ako $u, v ∈ V(G_k)$ to znači da su oni na rubu iste oblasti i dodavanjem grane $\{u, v\}$ dijeli tu oblast na dva dijela, to jest:
$$f_{k+1} = f_k + 1$$
$$|V_{k+1}| = |V_k|$$
$$|E_{k+1}| = |E_k| + 1$$

Koristeći induktivnu pretpostavku dobijamo
$$f_k = |E_k| - |V_k| + 2 ⇒ f_k + 1 = |E_k| + 1 - |V_k| + 2 ⇒ f_{k+1} = |E_{k+1}| - |V_{k+1}| + 2.$$

**$II)$** Ako $u ∈ V(G_k)$ i $v ∉ V(G_k)$ to znači da je novi čvor $v$ u unutrašnjosti neke oblasti u kojoj je čvor $u$ na rubu, što znači da uvođenjem grane $\{u, v\}$ se ne dijeli ta oblast na dva dijela, to jest:
$$f_{k+1} = f_k$$
$$|V_{k+1}| = |V_k| + 1$$
$$|E_{k+1}| = |E_k| + 1$$

Koristeći induktivnu pretpostavku dobijamo
$$f_k = |E_k| - |V_k| + 2 ⇒ f_k = (|E_k| + 1) - (|V_k| + 1) + 2 ⇒ f_k = |E_k| + 1 - |V_k| - 1 + 2 ⇒ f_{k+1} = |E_{k+1}| - |V_{k+1}| + 2$$

čime je dokaz završen.


## **Stepen oblasti**

#### **Definicija 2.** Stepen oblasti $R$, u oznaci $st(R)$ predstavlja broj grana na rubu te oblasti. Ako se grana pojavljuje dva puta na rubu oblasti ona se računa dva puta.

**ANALOGIJA:** Možemo zamisliti da je svaka grana podjeljena na dvije strane, i onda je stepen oblasti zapravo broj strana kroz koje grana "gleda" na tu oblast. Ako je grana na rubu oblasti, onda ona "gleda" tu oblast samo sa jednom svojom stranom, dok druga strana gleda na spoljašnjost oblasti, a ako je grana sva u unutrašnjosti oblasti ili ima jedan čvor koji nije na rubu, tada ona "gleda" tu oblast sa obje svoje strane. Strane su kao prozori.

To objašnjava zašto **graf koji ima tačno dva čvora i jednu granu određuje oblast koja ima stepen dva**, jer ta grana sa obje strane "gleda" tu oblast.

Takođe **ako u povezanom grafu postoje bar 3 čvora onda je svaka oblast u tom grafu bar stepena 3**.

##### **Primer 4.** Stepeni **oblasti**

image.png

Stepeni oblasti u primjeru su sledeći $R_1 = 3$, $R_2 = 6$ i $R_3 = 7.$

**VAŽNO:** Kod planarnog grafa $G = (V, E)$, njegova planarna reprezentacija dijeli ravan na oblasti $R_1$, $R_2$, $R_3$, ..., $R_l$. Pošto se svaka grana u zbiru stepena oblasti računa dva puta *(ili "gleda" sa obje strane u istu oblast, ili s jedne strane gleda u jednu a s druge u drugu oblast)* to znači da je

$$\sum_{1≤i≤l} st(R_i) = 2|E(G)|.$$

#### **Posledica 1.** Ako je graf $G = (V, E)$, $|V| ≥ 3$, **povezan planaran prost** graf i $f$ broj oblasti na koje on dijeli ravan, tada je $|E| ≤ 3⋅|V| - 6.$

**Dokaz.** S obzirom na to da je broj čvorova bar $3$, to znači da svaka oblast koju graf formira je bar stepena $3$. Znamo i da je suma svih stepena oblasti jednaka dvostrukom broju grana grafa, pa onda važi da je

$$2|E| = \sum_{1≤i≤f}st(R_i) ≥ 3 \cdot f ⇒ f ≤ \frac{2}{3}|E|.$$

Koristeći sada Ojlerovu formulu dobijamo

$$|E| - |V| + 2 ≤ \frac{2}{3}|E| ⇔ |E| ≤ 3⋅|V| - 6.$$


#### **Posledica 2.** Ako je graf $G = (V, E)$, $|V| ≥ 3$, **povezan planaran prost** graf bez kontura dužine $3$, tada je $|E| ≤ 2⋅|V| - 4.$

**Dokaz.** Pošto se u grafu ne nalaze konture dužine $3$, to znači da svaka oblast koju graf formira mora biti minimalno stepena $4$. Ako je $f$ broj oblasti koje graf formira, tada je

$$2⋅|E| = \sum_{1≤i≤f}st(R_i) ≥ 4⋅f ⇒ f ≤ \frac{1}{2}|E|.$$

Koristeći sada Ojlerovu formulu dobijamo

$$|E| - |V| + 2 ≤ \frac{1}{2}|E| ⇔ |E| ≤ 2⋅|V| - 4.$$

#### **Posledica 3.** Ako je graf $G = (V, E)$ **povezan planaran prost** graf, tada on sadrži čvor koji je stepena manjeg od šest.

**Dokaz.** Ako imamo graf koji ima jedan ili dva čvora, tvrđenje automatski važi. Razmatrajmo povezane planarne proste grafove koji sadrže bar 3 čvora.
Znamo da je

$$2⋅|E| = \sum_{v ∈ V(G)}deg(v).$$

Pretpostavimo suprotno, da tvrđenje ne važi, to jest da je svaki čvor bar stepena šest. To bi onda sledilo da je

$$2⋅|E| = \sum_{v ∈ V(G)}deg(v) ≥ 6⋅|V|.$$

**Posljedica 1** (čije uslove ispunjavamo) nalaže da je

$$|E| ≤ 3⋅|V| - 6 ⇔ 2⋅|E| ≤ 6⋅|V| - 12$$

Tako dobijamo da je

$$6⋅|V| ≤ 2⋅|E| ≤ 6⋅|V| - 12 ⇔ 0 ≤ -12,$$

što je očigledna kontradikcija, dakle polazno tvrđenje važi.

Sada slijede neki od primjera gdje se koriste posledice Ojlerove formule i stepena oblasti.

##### **Primer 5.** Pokazati da kompletan graf $K_5$ nije planaran.

**Dokaz.** Pretpostavimo suprotno, da $K_5$ jeste planaran. Kao takav nad njim mora da važi ***posljedica 1***, to jest da je

$$|E| ≤ 3⋅|V| - 6$$

U $K_5$ imamo da je $|E| = \binom{5}{2} = 15$, $|V| = 5$. Posledica očigledno ne važi jer

$$10 ≤ 15 - 6 = 9 ⇔ 10 ≤ 9$$

je netačno. Što znači da $K_5$ nije planaran.


##### **Primer 6.** Pokazati da kompletan bipartitan graf $K_{3,3}$ nije planaran.

**Dokaz.** Pretpostavimo suprotno, da jeste planaran. Pošto u sebi ne sadrži konturu dužine $3$ nad njim mora da važi ***posljedica 2***, to jest da je

$$|E| ≤ 2⋅|V| - 4$$

U $K_{3,3}$ imamo da je $|E| = 3⋅3 = 9$, $|V| = 6$. Posledica očigledno ne važi jer

$$9 ≤ 2⋅6 - 4 = 8 ⇔ 9 ≤ 8$$

je netačno. Što znači da $K_{3,3}$ nije planaran.


## **Teorema Kuratovskog**

Pokazavši da grafovi $K_5$ i $K_{3,3}$ nisu planarni možemo zaključiti da svaki graf koji sadrži ova dva grafa kao svoje podgrafove sigurno nije planaran. Možemo ići i korak dalje pa zaključiti da je svaki podgraf planarnog grafa takođe planaran, i ako bar jedan podgraf grafa nije planaran onda ni sam graf nije planaran, jer ne možemo ostvariti planarnu reprezentaciju cijelog grafa ako ne možemo planarno predstaviti svaki njegov podgraf.

#### **Definicija 3.** Elementarna deoba grane $\{u, v\}$ predstavlja umetanje novog čvora $w$ na tu granu, zatim uklanjanje te grane, tako od jedne grane formirajući dvije nove i to $\{u, w\}$ i $\{w, v\}.$

Ako u prostom planarnom grafu izvršimo elementarnu deobu grane $\{u, v\}$ dodavanjem čvora $w$ i zamenom grane $\{u, v\}$ sa novonastalim granama $\{u, w\}$ i $\{w, v\}$ graf ostaje planaran. Na ovaj način se uvek dodaje čvor stepena dva.

#### **Definicija 4.** Za grafove $G_1 = (V_1, E_1)$ i $G_2 = (V_2, E_2)$ kaže se da su **homeomorfni** ako se mogu dobiti od istog grafa primjenom konačno mnogo elementarnih deoba grane.

#### **Teorema 2. (Kuratovskog)** Graf $G = (V, E)$ je neplanaran ako i samo ako sadrži podgraf koji je homeomorfan sa $K_{3,3}$ ili $K_5.$

Slijedi par primera koji ilustruju ovu teoremu.

##### **Primer 7. (Petersenov graf)** Odredi da li je dati graf $G$ planaran.

**Rešenje.** Ako obrišemo čvor $b$ i uklonimo tri incidentne grane čvora $b$, dobijamo graf koji je homeomorfan sa grafom $K_{3,3}$, sa skupovima čvorova $\{f, d, j\}$ i $\{e, i, h\}$.

image.png

Do toga zaključka dolazimo preko niza elementarnih deoba grane, uklanjanjem $\{d, h\}$ i dodavanjem novonastalih grana $\{c, h\}$ i $\{c, d\}$, zatim uklanjanjem $\{e, f\}$ i dodavanjem novih grana $\{a, e\}$ i $\{a, f\}$ i konačno uklanjanjem $\{i, j\}$ i dodavanjem $\{g, i\}$ i $\{g, j\}$. Tako da graf $G$ prema teoremi Kuratovskog nije planaran, jer smo pronašli homeomorfan podgraf grafu $K_{3,3}$.

##### **Primer 8.** Koristeći teoremu Kuratovskog, odredi da li je dati graf planaran.

image.png

**Rešenje.** Prema teoremi Kuratovskog graf nije planaran ako i samo ako sadrži podgraf homeomorfan ili sa $K_5$ ili sa $K_{3,3}$, dakle određenim uklanjanjem grana i/ili čvorova bi trebali dobiti ili $K_5$ ili $K_{3,3}$. Ako posmatramo dati graf vidimo da sadrži $8$ čvorova kod koga je svaki čvor stepena $3$. To automatski eliminiše postojanje podgrafa homeomorfnog $K_5$, zato što je u $K_5$ minimalan stepen čvora $5$, dok je u našem grafu maksimalan stepen čvora $3$. Dakle ostaje da se provjeri da li postoji podgraf homeomorfan sa $K_{3,3}$. Da bi došli do $K_{3,3}$ potrebno je ukloniti $2$ čvora, a pošto je graf simetričan, odabir prvog čvora nije bitan. Ako izaberemo bilo koji čvor, i uklonimo ga, dobijamo grafovski multiskup $\{2, 2, 2, 3, 3, 3, 3\}$ što znači da već u prvom koraku vidimo da ne postoji ni podgraf homeomorfan grafu $K_{3,3}$, zato što imamo samo $4$ čvora stepena $3$, a potrebna nam je minimalno $6$ čvorova stepena $3$. Dakle ovaj graf jeste planaran, a na slici je data njegova planarna reprezentacija.
