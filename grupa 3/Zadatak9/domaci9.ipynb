## **Stablo**

**_Definicija_**:

> Stablo je neorijentisani graf koji zadovoljava dva uslova:
>
> - Povezanost: Za svaki par čvorova $u,v∈V(G)$, postoji bar jedan put između $u$ i $v$. Graf $G$ je topološki povezan.
> - Acikličnost: Graf $G$ ne sadrži nijedan ciklus, odnosno ne postoji zatvoren hod u kome se nijedna grana ne ponavlja, a početni i krajnji čvor su identični.

## **Šuma**

**_Definicija_**:

> Graf $G$ je šuma ako ne sadrži nijedan ciklus kao podgraf. Svaka povezana komponenta grafa $G$ koji je šuma, je stablo. Stoga se šuma može posmatrati kao disjunktna unija stabala.

_Primer:_

Primera šume može biti primer stabala iznad posmatran kao celina, **šuma**.


### **Ekvivalentne tvrdnje za stablo**

Neka je $G = (V, E)$ konačan, prost, neusmeren graf sa $|V| = n$. Sledećih pet tvrdnji su međusobno **ekvivalentne** i karakterišu **stablo**:

**_Tvrdnja 1 (Standardna definicija)_**

> Graf $G$ je stablo ako je **povezan i acikličan**.

**_Tvrdnja 2 (Minimalna povezanost)_**

> Graf $G$ je stablo ako je **minimalno povezan**, tj. ako je povezan i uklanjanje bilo koje grane čini graf nepovezanim.

**_Tvrdnja 3 (Jedinstveni putevi)_**

> Graf $G$ je stablo ako za svaki par različitih čvorova u $G$ postoji **tačno jedan prost put**.

**_Tvrdnja 4 (Dodavanje grane stvara konturu)_**

> Graf $G$ je stablo ako je **acikličan**, i ako dodavanje bilo koje nove grane između nepovezanih čvorova stvara **tačno jednu konturu**.

**_Tvrdnja 5 (Rekonstruktivna karakterizacija)_**

> Graf $G$ je stablo ako za svaki neprazan podgraf $H \subseteq G$ važi:
>
> $$
> \forall H \subseteq G, H \neq \emptyset: \quad |V(H)| = |E(H)| + c(H)
> $$
>
> gde je $c(H)$ broj komponenti povezanosti grafa $H$.  
> U slučaju kada je $G$ povezan ($c(G) = 1$), onda:$$
>|V| = |E| + 1 \Rightarrow |E| = n - 1$$

---

### **Leme**

U cilju dokazivanja međusobne ekvivalencije tvrdnji $(T1)-(T5)$, koristićemo sledeće formalne leme. U svakoj se pretpostavlja da je $G = (V, E)$ konačan, prost, neusmeren graf sa $n = |V|$ čvorova i $m = |E|$ grana.

#### **_Lema 1_**

> Ako je $G$ graf sa $n$ čvorova i $n - 1$ grana, i $G$ je povezan, tada je $G$ stablo.

_Dokaz:_  
Pretpostavimo da graf $G$ ima $n$ čvorova i tačno $n - 1$ grana, te da je povezan.  
Posmatrajmo da li je moguće da $G$ sadrži ciklus.

Neka postoji ciklus $C \subseteq G$. Tada možemo ukloniti jednu granu iz ciklusa $C$, čime se graf ne bi raskinuo (jer bi postojala alternativna putanja u preostalom delu ciklusa). Dakle, graf bi ostao povezan i imao bi $n - 2$ grana.

Ali iz poznate tvrdnje iz teorije grafova znamo:
Svaki povezan neusmeren graf sa $n$ čvorova mora imati **najmanje** $n - 1$ grana.

Ovde imamo manje od $n - 1$ grana, ali i dalje povezan graf — **kontradikcija**. Dakle, početna pretpostavka o postojanju ciklusa mora biti netačna.

Iz toga sledi da je $G$ acikličan.

Dakle, graf $G$ je:

- povezan (po pretpostavci)
- acikličan (dokazano)

što znači da zadovoljava definiciju stabla. $\square
$

---

##### **_Lema 2_**

> Ako je $G$ povezan i acikličan, tada je $G$ stablo.

_Dokaz:_  
Pretpostavimo da je $G$ povezan i da ne sadrži nijedan ciklus (acikličan).

Iz poznatog rezultata u teoriji grafova, za svaki acikličan graf (tj. **šumu**) važi da:

$$
|E| \leq |V| - c
$$

gde je $c$ broj povezanih komponenti.

Pošto je $G$ povezan, imamo $c = 1$, pa sledi:

$$
|E| \leq |V| - 1 = n - 1
$$

Ali po pretpostavci je $G$ povezan, što znači da mora sadržati **barem $n - 1$** grana da bi svi čvorovi bili međusobno dostižni. Dakle:

$$
|E| \geq n - 1
$$

Kombinovanjem dobijamo:

$$
|E| = n - 1
$$

Dakle, $G$ je povezan, acikličan i ima tačno $n - 1$ granu, što prema standardnoj definiciji karakteriše stablo. $\square
$

---

#### **_Lema 3_**

> Ako između svakog para čvorova u $G$ postoji tačno jedan prost put, tada je $G$ stablo.

_Dokaz:_  
Neka je $G$ takav da za svaka dva različita čvora $u, v \in V$ postoji tačno jedan prost put $P_{uv}$.

To implicira sledeće:

- **Povezanost**: Za svaka dva čvora postoji put, dakle $G$ je povezan.
- **Jedinstvenost**: Za svaka dva čvora postoji najviše jedan prost put $\Rightarrow$ ne može postojati nijedan ciklus.

Pretpostavimo suprotno, tj. da $G$ sadrži ciklus $C$. Tada možemo pronaći bar dva različita puta između nekih parova čvorova u $C$ koristeći dve različite grane konture. To bi bilo u suprotnosti sa uslovom da postoji tačno jedan prost put između svakog para čvorova. Dakle, postojanje ciklusa implicira višestruke puteve — kontradikcija.

Prema tome, $G$ ne može imati ciklus, tj. acikličan je.

Pošto je $G$:

- povezan
- acikličan

zaključujemo da je $G$ stablo. $\square$

---

#### **_Lema 4_**

> Ako je $G$ acikličan, i dodavanje bilo koje grane između nepovezanih čvorova stvara tačno jednu konturu, tada je $G$ stablo.

_Dokaz:_  
Pretpostavimo da je $G$ acikličan i da za svaki par čvorova $u, v \in V$, ako $uv \notin E$, tada dodavanje grane $uv$ formira **tačno jedan ciklus**.

Analizirajmo povezanost:

Pretpostavimo suprotno, tj. da $G$ **nije povezan**. Tada postoji barem jedan par čvorova $u, v$ između kojih ne postoji put u $G$.

U tom slučaju, dodavanjem grane $uv$ **ne može nastati kontura**, jer $u$ i $v$ pripadaju različitim komponentama i dodavanjem grane se formira **prvi put** između njih, a ne ciklus.

To je u kontradikciji sa pretpostavkom da svako dodavanje nove grane zatvara jedinstvenu konturu.

Dakle, $G$ mora biti povezan.

Pošto je već po pretpostavci acikličan, zaključujemo:

- $G$ je acikličan
- $G$ je povezan

$\Rightarrow$ $G$ je stablo. $\square$

---

**_Teorema 1_**

> **Prost graf sa $n$ čvorova i $n - 1$ grana je stablo ako i samo ako je povezan.**

**Dokaz:**

- $(\Rightarrow)$ Po definiciji.
- $(\Leftarrow)$ Ako je graf povezan i ima $n - 1$ grana, mora biti bez ciklusa $\Rightarrow$ stablo.

**Teorema 2**

> **Graf $G$ je stablo ako i samo ako za svaki neprazan podgraf $H \subseteq G$ važi: $|V(H)| = |E(H)| + 1$.**

**Dokaz:**

- Ova osobina je poznata karakterizacija stabla pomoću topoloških invarijanti.
- Važi i za ceo graf i za sve njegove podgrafove.
- U suprotnom bi postojala komponenta sa viškom grana (ciklus) ili nedostatkom (nepovezanost).

---

Na osnovu dokazanih lema i teorema, zaključujemo:

$$
\text{Graf } G \text{ je stablo} \iff \text{(T1)} \iff \text{(T2)} \iff \text{(T3)} \iff \text{(T4)} \iff \text{(T5)}
$$


## **Priferov kod**

Priferov kod je jedan od najvažnijih alata za **kodiranje stabala** u teoriji grafova. Ime je dobio po nemačkom matematičaru Heinz Prüferu koji ga je uveo 1918. godine.

Ova metoda kodira **etiketirano stablo** sa $n$ čvorova u **niz od $n - 2$ celih brojeva**, pri čemu svaki broj u nizu pripada skupu $\{1, 2, \dots, n\}$. Proces je **bijektivna transformacija**, što znači da postoji **jedinstveno stablo** za svaki Priferov kod i obrnuto.

**_Definicija_**:

Neka je $T$ etiketirano stablo sa skupom čvorova $V = \{1, 2, \dots, n\}$.

**Priferov kod** stabla $T$ je niz $(a_1, a_2, \dots, a_{n-2})$ koji se dobija sledećim algoritmom:

---

### **Algoritam za generisanje Priferovog koda iz stabla**

1. Dok god u stablu postoji više od dva čvora:
   - Pronađi čvor sa **najmanjim elementom** koji ima stepen 1, to jest list.
   - Zapiši njegovog suseda (roditelja) u Priferov niz.
   - Ukloni taj list iz stabla (zajedno sa pripadajućom granom).
2. Nastavi dok ne ostanu dva čvora — oni se ne uključuju u kod.

Na kraju se dobija niz od tačno $n - 2$ elemenata.

```python
import networkx as nx

def prufer_encode(tree_edges):
    """
    Prima: listu grana stabla (npr. [(1,2), (1,3), (3,4), ...])
    Vraća: Priferov kod kao listu celih brojeva
    """
    import heapq
    from collections import defaultdict

    # Skup svih čvorova
    nodes = set()
    for u, v in tree_edges:
        nodes.add(u)
        nodes.add(v)

    n = len(nodes)
    degree = defaultdict(int)
    graph = defaultdict(set)

    for u, v in tree_edges:
        degree[u] += 1
        degree[v] += 1
        graph[u].add(v)
        graph[v].add(u)

    # Min-heap svih listova (čvorovi sa stepenom 1)
    leaves = [node for node in nodes if degree[node] == 1]
    heapq.heapify(leaves)

    code = []
    for _ in range(n - 2):
        leaf = heapq.heappop(leaves)
        neighbor = next(iter(graph[leaf]))

        code.append(neighbor)

        degree[neighbor] -= 1
        degree[leaf] -= 1

        graph[neighbor].remove(leaf)
        graph[leaf].remove(neighbor)

        if degree[neighbor] == 1:
            heapq.heappush(leaves, neighbor)

    return code

# Primer poziva:
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (2, 6), (6, 7)]
prufer_code = prufer_encode(edges)
print("Priferov kod:", prufer_code)
```

**_Napomena:_**  
Za neorijentisani graf $G=(V,E)$, i čvor $v∈V$, broj ponavljanja etikete čvora u Priferovom kodu jednak je $deg(v) - 1$ . U ovom primeru, čvor $2$ se pojavljuje 3 puta jer $deg(2) = 4$ (povezan je sa čvorovima 1, 4, 5, i 6, posredno). Ovo je dodatna provera doslednosti algoritma.

**_Teorema:_**

Broj etiketiranih stabala

Za dati broj $n$ čvorova, broj različitih etiketiranih stabala jednak je:

$$
n^{n - 2}
$$

Ova formula je poznata kao **Cayleyjeva formula**, a direktna posledica činjenice da postoji bijekcija između etiketiranih stabala i Priferovih kodova dužine $n - 2$ sa elementima iz $\{1, 2, \dots, n\}$.


---

### **Algoritam za rekonstrukcija stabla iz Priferovog koda**

Neka je dat Priferov kod:

$$
P = (a_1, a_2, \dots, a_{n-2})
$$

gde je svaki $a_i \in \{1, 2, \dots, n\}$.

Koraci za rekonstrukciju etiketiranog stabla sa $n$ čvorova su sledeći:

1. Inicijalizuj niz $D$ dužine $n$, gde je $D[v]$ broj pojavljivanja čvora $v$ u Priferovom kodu $P$.
2. Za sve čvorove $v \in \{1, \dots, n\}$ koji **ne postoje** u $P$, postavi $D[v] = 0$.
3. Napravi pomoćnu strukturu (npr. prioritetni red) da efikasno čuva čvorove sa $D[v] = 0$ — to su trenutno **listovi**.
4. Za svaki element $a_i$ u kodu, redom:
   - Uzmi **najmanji čvor $v$** takav da je $D[v] = 0$ (najmanji dostupni list).
   - Dodaj granu $(v, a_i)$.
   - Ažuriraj $D[a_i] := D[a_i] - 1$, a $D[v] := -1$ (ili ga ukloni iz razmatranja).
   - Ako nakon umanjenja $D[a_i]$ postane 0, dodaj $a_i$ u skup listova.
5. Kada obradiš sve elemente niza, ostaju još **dva čvora sa $D[v] = 0$** — poveži ih poslednjom granom.

Na kraju dobijaš povezano stablo sa $n$ čvorova i $n - 1$ grana.


_Primer_ :

**Rekonstrukcija stabla iz Priferovog koda**

Neka je dat Priferov kod:

$$
P = (1, 2, 2, 2, 6)
$$

**Korak 1:** Odredi broj čvorova

Priferov kod dužine $n - 2 = 5$ implicira da je broj čvorova:

$$
n = 7
$$

Dakle, skup čvorova je:

$$
V = \{1, 2, 3, 4, 5, 6, 7\}
$$

**Korak 2:** Inicijalizuj broj pojavljivanja (frekvencije)

| Čvor | Broj pojavljivanja u kodu | Inicijalni stepen |
| ---- | ------------------------- | ----------------- |
| 1    | 1                         | 2                 |
| 2    | 3                         | 4                 |
| 3    | 0                         | 1                 |
| 4    | 0                         | 1                 |
| 5    | 0                         | 1                 |
| 6    | 0                         | 1                 |
| 7    | 1                         | 2                 |

(Stepen = broj pojavljivanja + 1)

**Korak 3:** Iterativno formiraj grane

Inicijalni skup listova (čvorovi sa $0$ pojavljivanja): $\{3, 4, 5, 6\}$  
Sortirano: $[3, 4, 5, 6]$

Obrađujemo kod redom: $(1, 2, 2, 2, 6)$

- **Prvi element:** $1$  
  Najmanji list: $3$ ⇒ grana $(3, 1)$  
  Ažuriramo: $D[3] = -1$, $D[1] = 0$ ⇒ dodajemo $1$ u listove

- **Drugi element:** $2$  
  Najmanji list: $1$ ⇒ grana $(1, 2)$  
  Ažuriramo: $D[1] = -1$, $D[2] = 2$

- **Treći element:** $2$  
  Najmanji list: $4$ ⇒ grana $(4, 2)$  
  Ažuriramo: $D[4] = -1$, $D[2] = 1$

- **Četvrti element:** $2$  
  Najmanji list: $5$ ⇒ grana $(5, 2)$  
  Ažuriramo: $D[5] = -1$, $D[2] = 0$ ⇒ dodajemo $2$ u listove

- **Peti element:** $6$  
  Najmanji list: $2$ ⇒ grana $(2, 6)$  
  Ažuriramo: $D[2] = -1$, $D[6] = 0$ ⇒ dodajemo $6$ u listove

Na kraju ostaju čvorovi sa $D[v] = 0$: čvorovi $6$ i $7$  
⇒ dodajemo poslednju granu $(6, 7)$

**Dobijeno stablo**

Skup grana rekonstruisanog stabla:

$$
\{(3,1), (1,2), (4,2), (5,2), (2,6), (6,7)\}
$$

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict

def prufer_decode(prufer_code):
    """
    Prima: Priferov kod kao listu celih brojeva
    Vraća: listu grana rekonstruisanog stabla
    """
    n = len(prufer_code) + 2
    degree = defaultdict(int)

    for node in prufer_code:
        degree[node] += 1

    # Inicijalni čvorovi
    nodes = set(range(1, n + 1))
    result = []

    for node in prufer_code:
        for leaf in sorted(nodes):
            if degree[leaf] == 0:
                result.append((leaf, node))
                degree[node] -= 1
                nodes.remove(leaf)
                break

    # Preostala dva čvora
    u, v = sorted(nodes)
    result.append((u, v))

    return result

def hierarchy_pos(G, root=None, width=1.0, vert_gap=0.2, vert_loc=0, xcenter=0.5, pos=None, parent=None):
    """
    Hijerarhijski layout za stablo.
    """
    if pos is None:
        pos = {root: (xcenter, vert_loc)}
    else:
        pos[root] = (xcenter, vert_loc)
    children = list(G.neighbors(root))
    if parent is not None:   # ukloni roditelja iz liste dece
        children.remove(parent)
    if len(children) != 0:
        dx = width / len(children)
        next_x = xcenter - width/2 - dx/2
        for child in children:
            next_x += dx
            pos = hierarchy_pos(G, root=child, width=dx, vert_gap=vert_gap,
                                vert_loc=vert_loc - vert_gap, xcenter=next_x, pos=pos, parent=root)
    return pos


def plot_tree(edges):
    """
    Iscrtava stablo zadato listom grana u hijerarhijskom rasporedu.
    """
    import matplotlib.pyplot as plt
    import networkx as nx

    G = nx.Graph()
    G.add_edges_from(edges)

    degree = dict(G.degree())
    root = min([node for node, deg in degree.items() if deg > 1], default=list(G.nodes)[0])

    pos = hierarchy_pos(G, root)

    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color='lightgreen',
            node_size=1000, font_size=14, edge_color='gray')
    plt.title("Prikaz stabla", fontsize=16)
    plt.show()


# Primer poziva:
code = [1, 2, 2, 2, 6]
edges = prufer_decode(code)
print("Rekonstruisane grane:", edges)
plot_tree(edges)
```

## **Težinski grafovi i minimalna pokrivajuća stabla**

### **Težinski graf**

**_Definicija:_**

> Neka je $G = (V, E)$ konačan, neusmeren graf. Ako je svakoj grani $e \in E$ dodeljena realna vrednost $w(e) \in \mathbb{R}$ (koja predstavlja "trošak", "dužinu", "težinu" ili sličan atribut), tada kažemo da je $G$ **težinski graf**.
> Formalno, imamo funkciju težine:

$$
w : E \rightarrow \mathbb{R}
$$

U kontekstu usmerenih grafova, težine se dodeljuju **lukovima**, tj. parovima $(u, v)$, što omogućava i različite vrednosti $w(u, v) \neq w(v, u)$.

### **Minimalno pokrivajuće stablo (MST)**

**_Definicija_**:

> Za povezani težinski neusmeren graf $G = (V, E)$, **minimalno pokrivajuće stablo** (engl. _Minimum Spanning Tree_, skraćeno **MST**) je podgraf $T = (V, E')$ takav da:
> - $T$ je stablo: povezan i acikličan,
>- $E' \subseteq E$, i $|E'| = |V| - 1$,
>- Ukupna težina je minimalna: $$\sum_{e \in E'} w(e) = \min$$

MST ne mora biti jedinstven — može postojati više različitih stabala iste minimalne težine.


## **Algoritmi za određivanje minimalnog pokrivajućeg stabla**

---

### **1. Kruskalov algoritam**

**Ideja:**

Postupno dodajemo najjeftinije grane koje ne formiraju ciklus, sve dok ne dobijemo stablo sa $n - 1$ grana.

**Koraci:**

1. Sortiraj sve grane rastuće po težini.
2. Inicijalizuj $T = \emptyset$ (nema grana).
3. Za svaku granu $e = (u, v)$ iz sortirane liste:
   - Ako dodavanje $e$ u $T$ **ne stvara ciklus** ⇒ dodaj $e$ u $T$.
4. Zaustavi se kada $|T| = n - 1$.

**Struktura podataka:**

Za detekciju ciklusa koristi se **Union-Find (Disjoint Set Union, DSU)** struktura sa optimizacijama:

- spajanje po rangu
- putna kompresija

```python
import networkx as nx

def kruskal_mst(G):
    """
    Kruskalov algoritam za pronalaženje minimalnog pokrivajućeg stabla (MST).
    Ulaz: neusmeren težinski graf G (networkx.Graph)
    Izlaz: skup grana MST kao lista (u, v, težina)
    """
    # Inicijalizuj disjunktne skupove (Union-Find)
    parent = {}
    def find(u):
        while parent[u] != u:
            parent[u] = parent[parent[u]]
            u = parent[u]
        return u
    def union(u, v):
        ru, rv = find(u), find(v)
        if ru != rv:
            parent[ru] = rv
            return True
        return False

    for node in G.nodes:
        parent[node] = node

    mst_edges = []
    sorted_edges = sorted(G.edges(data=True), key=lambda e: e[2].get('weight', 1))

    for u, v, data in sorted_edges:
        if union(u, v):
            mst_edges.append((u, v, data.get('weight', 1)))
            if len(mst_edges) == G.number_of_nodes() - 1:
                break

    return mst_edges

# Primer poziva:
G = nx.Graph()
G.add_weighted_edges_from([
    (1, 2, 4),
    (1, 3, 5),
    (2, 3, 1),
    (3, 4, 3),
    (2, 4, 2)
])

print("Kruskal MST:", kruskal_mst(G))
```

### **2. Primov algoritam**

**Ideja:**

Postupno izgrađujemo stablo tako što u svakoj iteraciji dodajemo najjeftiniju granu koja povezuje već izabrani deo grafa sa nekim neizabranim čvorom.

**Koraci:**

1. Odaberi proizvoljan početni čvor $s \in V$.
2. Inicijalizuj $T = \{s\}$, a skup $E_T = \emptyset$.
3. Dok $T \neq V$:
   - Nađi granu $(u, v)$ sa najmanjom težinom takvu da je $u \in T$ i $v \notin T$.
   - Dodaj čvor $v$ i granu $(u,v)$ u $T$.

**Struktura podataka:**

Efikasna implementacija koristi **min-heap (prioritetni red)** za održavanje najjeftinijih kandidata za proširenje.

![Vizualizacija Primovog algoritma](https://upload.wikimedia.org/wikipedia/commons/9/9b/PrimAlgDemo.gif)


```python
import heapq

def prim_mst(G, start=None):
    """
    Primov algoritam za minimalno pokrivajuće stablo (MST).
    Ulaz: težinski neusmeren graf G (networkx.Graph)
    start: početni čvor (ako nije dat, koristi najmanji čvor)
    Izlaz: lista grana (u, v, weight)
    """
    if start is None:
        start = list(G.nodes)[0]

    visited = set()
    mst_edges = []
    min_heap = []

    def add_edges(node):
        visited.add(node)
        for neighbor, data in G[node].items():
            if neighbor not in visited:
                weight = data.get('weight', 1)
                heapq.heappush(min_heap, (weight, node, neighbor))

    add_edges(start)

    while min_heap and len(visited) < G.number_of_nodes():
        weight, u, v = heapq.heappop(min_heap)
        if v not in visited:
            mst_edges.append((u, v, weight))
            add_edges(v)

    return mst_edges

# Primer poziva:
G = nx.Graph()
G.add_weighted_edges_from([
    (1, 2, 4),
    (1, 3, 5),
    (2, 3, 1),
    (3, 4, 3),
    (2, 4, 2)
])
print("Prim MST:", prim_mst(G, start=1))

```

## **Najkraći putevi u grafu**

**_Definicija:_**

> Za težinski **usmeren** ili **neusmeren** graf $G = (V, E)$ sa funkcijom težine $w : E \to \mathbb{R}_{\geq 0}$, **najkraći put** između čvorova $s$ i $v$ je put $P = (s, \dots, v)$ koji minimizuje:

$$
w(P) = \sum_{e \in P} w(e)
$$


### **Dijkstrin algoritam**

**Uslov:**

Sve težine u grafu moraju biti **nenegativne**: $w(e) \geq 0$ za sve $e \in E$.

**Ulaz:**

- Graf $G = (V, E)$
- Izvorni čvor $s \in V$

**Izlaz:**

- Niz $d[v]$ — najkraća rastojanja od $s$ do svakog $v \in V$

**Koraci algoritma:**

1. Inicijalizuj rastojanja:
   $$
   d[v] :=
   \begin{cases}
     0, & \text{ako } v = s \\
     \infty, & \text{inače}
   \end{cases}
   $$
2. Inicijalizuj prioritetni red $Q$ sa svim čvorovima (po $d[v]$).
3. Dok $Q$ nije prazan:
   - Ekstrahuj čvor $u$ sa najmanjom vrednošću $d[u]$ iz $Q$.
   - Za svaku susednu granu $(u, v)$:
     - Ako $d[v] > d[u] + w(u,v)$:
       - Ažuriraj $d[v] := d[u] + w(u,v)$
       - Ažuriraj poziciju $v$ u $Q$

**Složenost:**

- $O((|V| + |E|) \log |V|)$ uz **Fibonacci heap** (ili $O(|E| \log |V|)$ uz **binary heap**)

![Vizualizacija Dijkstrinog algoritma](https://upload.wikimedia.org/wikipedia/commons/5/57/Dijkstra_Animation.gif)



```python
import heapq

def dijkstra_shortest_path(G, start):
    """
    Dijkstrin algoritam za najkraće puteve (nenegativne težine).
    Ulaz: G - težinski graf (networkx), start - početni čvor
    Izlaz: rastojanja od start do svih drugih čvorova
    """
    dist = {node: float('inf') for node in G.nodes}
    dist[start] = 0

    visited = set()
    heap = [(0, start)]

    while heap:
        cost, u = heapq.heappop(heap)
        if u in visited:
            continue
        visited.add(u)

        for v, data in G[u].items():
            weight = data.get('weight', 1)
            if dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight
                heapq.heappush(heap, (dist[v], v))

    return dist

# Primer:
G = nx.DiGraph()
G.add_weighted_edges_from([
    (1, 2, 6),
    (1, 3, 5),
    (1, 4, 5),
    (2, 5, 1),
    (3, 2, 2),
    (3, 5, 1),
    (4, 3, 2),
    (4, 6, 1),
    (5, 7, 3),
    (6, 7, 3)
])
print(dijkstra_shortest_path(G, 1))
```

### **Bellman-Ford algoritam za najkraći put u grafu**

Bellman-Ford algoritam rešava problem najkraćeg puta iz jednog izvora $s$ ka svim ostalim čvorovima u težinskom grafu $G = (V, E)$, čak i kada **neke težine mogu biti negativne**.

Za razliku od Dijkstrinog algoritma, koji zahteva da sve težine budu nenegativne, Bellman-Ford funkcioniše i kada:

$$
\exists e \in E \text{ tako da } w(e) < 0
$$

**Ulaz:**

- Usmeren ili neusmeren težinski graf $G = (V, E)$ (može imati negativne težine)
- Početni čvor $s \in V$

**Izlaz:**

- Niz $d[v]$ najkraćih rastojanja od $s$ do svakog $v \in V$
- Ako postoji negativan ciklus koji je dostižan iz $s$, algoritam to detektuje

**Koraci algoritma:**

1. Inicijalizuj rastojanja:
   $$
   d[v] :=
   \begin{cases}
   0, & v = s \\
   \infty, & \text{inače}
   \end{cases}
   $$
2. **Ponovi $|V| - 1$ puta:**
   - Za svaku granu $(u, v)$:
     - Ako $d[u] + w(u, v) < d[v]$:
       - ažuriraj $d[v] := d[u] + w(u, v)$
3. **Detekcija negativnog ciklusa:**
   - Ako i nakon $|V| - 1$ iteracija postoji grana $(u, v)$ koja može ažurirati $d[v]$, tada postoji **negativni ciklus**

**Složenost:**

- Vremenska: $O(|V||E|)$
- Prostorna: $O(|V|)$

**Korišćenje:**

- Kada graf može imati negativne težine
- Kada je potrebno detektovati negativne cikluse
- U mrežnim algoritmima, dinamičkom programiranju, teoriji valova itd.

![Vizualizacija Bellman-Ford algoritma](https://upload.wikimedia.org/wikipedia/commons/7/77/Bellman%E2%80%93Ford_algorithm_example.gif)



```python
def bellman_ford_shortest_path(G, start):
    """
    Bellman-Ford algoritam za najkraće puteve (dozvoljene negativne težine).
    Ulaz: G - težinski usmeren graf (networkx.DiGraph), start - početni čvor
    Izlaz: dict rastojanja, ili Exception ako postoji negativan ciklus
    """
    dist = {node: float('inf') for node in G.nodes}
    dist[start] = 0

    # |V| - 1 iteracija
    for _ in range(len(G.nodes) - 1):
        for u, v, data in G.edges(data=True):
            weight = data.get('weight', 1)
            if dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight

    # Detekcija negativnog ciklusa
    for u, v, data in G.edges(data=True):
        weight = data.get('weight', 1)
        if dist[u] + weight < dist[v]:
            raise ValueError("Graf sadrži negativan ciklus.")

    return dist

# Primer:
G = nx.DiGraph()
G.add_weighted_edges_from([
    (1, 2, 6),
    (1, 3, 5),
    (1, 4, 5),
    (2, 5, -1),
    (3, 2, -2),
    (3, 5, 1),
    (4, 3, -2),
    (4, 6, -1),
    (5, 7, 3),
    (6, 7, 3)
])
print(bellman_ford_shortest_path(G, 1))
```
