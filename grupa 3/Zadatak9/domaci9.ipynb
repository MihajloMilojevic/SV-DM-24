## **Stablo**

**_Definicija_**:

> Stablo je neorijentisani graf koji zadovoljava dva uslova:
>
> - Povezanost: Za svaki par ƒçvorova $u,v‚ààV(G)$, postoji bar jedan put izmeƒëu $u$ i $v$. Graf $G$ je topolo≈°ki povezan.
> - Acikliƒçnost: Graf $G$ ne sadr≈æi nijedan ciklus, odnosno ne postoji zatvoren hod u kome se nijedna grana ne ponavlja, a poƒçetni i krajnji ƒçvor su identiƒçni.

## **≈†uma**

**_Definicija_**:

> Graf $G$ je ≈°uma ako ne sadr≈æi nijedan ciklus kao podgraf. Svaka povezana komponenta grafa $G$ koji je ≈°uma, je stablo. Stoga se ≈°uma mo≈æe posmatrati kao disjunktna unija stabala.

_Primer:_

Primera ≈°ume mo≈æe biti primer stabala iznad posmatran kao celina, **≈°uma**.


### **Ekvivalentne tvrdnje za stablo**

Neka je $G = (V, E)$ konaƒçan, prost, neusmeren graf sa $|V| = n$. Sledeƒáih pet tvrdnji su meƒëusobno **ekvivalentne** i karakteri≈°u **stablo**:

**_Tvrdnja 1 (Standardna definicija)_**

> Graf $G$ je stablo ako je **povezan i acikliƒçan**.

**_Tvrdnja 2 (Minimalna povezanost)_**

> Graf $G$ je stablo ako je **minimalno povezan**, tj. ako je povezan i uklanjanje bilo koje grane ƒçini graf nepovezanim.

**_Tvrdnja 3 (Jedinstveni putevi)_**

> Graf $G$ je stablo ako za svaki par razliƒçitih ƒçvorova u $G$ postoji **taƒçno jedan prost put**.

**_Tvrdnja 4 (Dodavanje grane stvara konturu)_**

> Graf $G$ je stablo ako je **acikliƒçan**, i ako dodavanje bilo koje nove grane izmeƒëu nepovezanih ƒçvorova stvara **taƒçno jednu konturu**.

**_Tvrdnja 5 (Rekonstruktivna karakterizacija)_**

> Graf $G$ je stablo ako za svaki neprazan podgraf $H \subseteq G$ va≈æi:
>
> $$
> \forall H \subseteq G, H \neq \emptyset: \quad |V(H)| = |E(H)| + c(H)
> $$
>
> gde je $c(H)$ broj komponenti povezanosti grafa $H$.  
> U sluƒçaju kada je $G$ povezan ($c(G) = 1$), onda:$$
>|V| = |E| + 1 \Rightarrow |E| = n - 1$$

---

### **Leme**

U cilju dokazivanja meƒëusobne ekvivalencije tvrdnji $(T1)-(T5)$, koristiƒáemo sledeƒáe formalne leme. U svakoj se pretpostavlja da je $G = (V, E)$ konaƒçan, prost, neusmeren graf sa $n = |V|$ ƒçvorova i $m = |E|$ grana.

#### **_Lema 1_**

> Ako je $G$ graf sa $n$ ƒçvorova i $n - 1$ grana, i $G$ je povezan, tada je $G$ stablo.

_Dokaz:_  
Pretpostavimo da graf $G$ ima $n$ ƒçvorova i taƒçno $n - 1$ grana, te da je povezan.  
Posmatrajmo da li je moguƒáe da $G$ sadr≈æi ciklus.

Neka postoji ciklus $C \subseteq G$. Tada mo≈æemo ukloniti jednu granu iz ciklusa $C$, ƒçime se graf ne bi raskinuo (jer bi postojala alternativna putanja u preostalom delu ciklusa). Dakle, graf bi ostao povezan i imao bi $n - 2$ grana.

Ali iz poznate tvrdnje iz teorije grafova znamo:
Svaki povezan neusmeren graf sa $n$ ƒçvorova mora imati **najmanje** $n - 1$ grana.

Ovde imamo manje od $n - 1$ grana, ali i dalje povezan graf ‚Äî **kontradikcija**. Dakle, poƒçetna pretpostavka o postojanju ciklusa mora biti netaƒçna.

Iz toga sledi da je $G$ acikliƒçan.

Dakle, graf $G$ je:

- povezan (po pretpostavci)
- acikliƒçan (dokazano)

≈°to znaƒçi da zadovoljava definiciju stabla. $\square
$

---

##### **_Lema 2_**

> Ako je $G$ povezan i acikliƒçan, tada je $G$ stablo.

_Dokaz:_  
Pretpostavimo da je $G$ povezan i da ne sadr≈æi nijedan ciklus (acikliƒçan).

Iz poznatog rezultata u teoriji grafova, za svaki acikliƒçan graf (tj. **≈°umu**) va≈æi da:

$$
|E| \leq |V| - c
$$

gde je $c$ broj povezanih komponenti.

Po≈°to je $G$ povezan, imamo $c = 1$, pa sledi:

$$
|E| \leq |V| - 1 = n - 1
$$

Ali po pretpostavci je $G$ povezan, ≈°to znaƒçi da mora sadr≈æati **barem $n - 1$** grana da bi svi ƒçvorovi bili meƒëusobno dosti≈æni. Dakle:

$$
|E| \geq n - 1
$$

Kombinovanjem dobijamo:

$$
|E| = n - 1
$$

Dakle, $G$ je povezan, acikliƒçan i ima taƒçno $n - 1$ granu, ≈°to prema standardnoj definiciji karakteri≈°e stablo. $\square
$

---

#### **_Lema 3_**

> Ako izmeƒëu svakog para ƒçvorova u $G$ postoji taƒçno jedan prost put, tada je $G$ stablo.

_Dokaz:_  
Neka je $G$ takav da za svaka dva razliƒçita ƒçvora $u, v \in V$ postoji taƒçno jedan prost put $P_{uv}$.

To implicira sledeƒáe:

- **Povezanost**: Za svaka dva ƒçvora postoji put, dakle $G$ je povezan.
- **Jedinstvenost**: Za svaka dva ƒçvora postoji najvi≈°e jedan prost put $\Rightarrow$ ne mo≈æe postojati nijedan ciklus.

Pretpostavimo suprotno, tj. da $G$ sadr≈æi ciklus $C$. Tada mo≈æemo pronaƒái bar dva razliƒçita puta izmeƒëu nekih parova ƒçvorova u $C$ koristeƒái dve razliƒçite grane konture. To bi bilo u suprotnosti sa uslovom da postoji taƒçno jedan prost put izmeƒëu svakog para ƒçvorova. Dakle, postojanje ciklusa implicira vi≈°estruke puteve ‚Äî kontradikcija.

Prema tome, $G$ ne mo≈æe imati ciklus, tj. acikliƒçan je.

Po≈°to je $G$:

- povezan
- acikliƒçan

zakljuƒçujemo da je $G$ stablo. $\square$

---

#### **_Lema 4_**

> Ako je $G$ acikliƒçan, i dodavanje bilo koje grane izmeƒëu nepovezanih ƒçvorova stvara taƒçno jednu konturu, tada je $G$ stablo.

_Dokaz:_  
Pretpostavimo da je $G$ acikliƒçan i da za svaki par ƒçvorova $u, v \in V$, ako $uv \notin E$, tada dodavanje grane $uv$ formira **taƒçno jedan ciklus**.

Analizirajmo povezanost:

Pretpostavimo suprotno, tj. da $G$ **nije povezan**. Tada postoji barem jedan par ƒçvorova $u, v$ izmeƒëu kojih ne postoji put u $G$.

U tom sluƒçaju, dodavanjem grane $uv$ **ne mo≈æe nastati kontura**, jer $u$ i $v$ pripadaju razliƒçitim komponentama i dodavanjem grane se formira **prvi put** izmeƒëu njih, a ne ciklus.

To je u kontradikciji sa pretpostavkom da svako dodavanje nove grane zatvara jedinstvenu konturu.

Dakle, $G$ mora biti povezan.

Po≈°to je veƒá po pretpostavci acikliƒçan, zakljuƒçujemo:

- $G$ je acikliƒçan
- $G$ je povezan

$\Rightarrow$ $G$ je stablo. $\square$

---

**_Teorema 1_**

> **Prost graf sa $n$ ƒçvorova i $n - 1$ grana je stablo ako i samo ako je povezan.**

**Dokaz:**

- $(\Rightarrow)$ Po definiciji.
- $(\Leftarrow)$ Ako je graf povezan i ima $n - 1$ grana, mora biti bez ciklusa $\Rightarrow$ stablo.

**Teorema 2**

> **Graf $G$ je stablo ako i samo ako za svaki neprazan podgraf $H \subseteq G$ va≈æi: $|V(H)| = |E(H)| + 1$.**

**Dokaz:**

- Ova osobina je poznata karakterizacija stabla pomoƒáu topolo≈°kih invarijanti.
- Va≈æi i za ceo graf i za sve njegove podgrafove.
- U suprotnom bi postojala komponenta sa vi≈°kom grana (ciklus) ili nedostatkom (nepovezanost).

---

Na osnovu dokazanih lema i teorema, zakljuƒçujemo:

$$
\text{Graf } G \text{ je stablo} \iff \text{(T1)} \iff \text{(T2)} \iff \text{(T3)} \iff \text{(T4)} \iff \text{(T5)}
$$


## **Priferov kod**

Priferov kod je jedan od najva≈ænijih alata za **kodiranje stabala** u teoriji grafova. Ime je dobio po nemaƒçkom matematiƒçaru Heinz Pr√ºferu koji ga je uveo 1918. godine.

Ova metoda kodira **etiketirano stablo** sa $n$ ƒçvorova u **niz od $n - 2$ celih brojeva**, pri ƒçemu svaki broj u nizu pripada skupu $\{1, 2, \dots, n\}$. Proces je **bijektivna transformacija**, ≈°to znaƒçi da postoji **jedinstveno stablo** za svaki Priferov kod i obrnuto.

**_Definicija_**:

Neka je $T$ etiketirano stablo sa skupom ƒçvorova $V = \{1, 2, \dots, n\}$.

**Priferov kod** stabla $T$ je niz $(a_1, a_2, \dots, a_{n-2})$ koji se dobija sledeƒáim algoritmom:

---

### **Algoritam za generisanje Priferovog koda iz stabla**

1. Dok god u stablu postoji vi≈°e od dva ƒçvora:
   - Pronaƒëi ƒçvor sa **najmanjim elementom** koji ima stepen 1, to jest list.
   - Zapi≈°i njegovog suseda (roditelja) u Priferov niz.
   - Ukloni taj list iz stabla (zajedno sa pripadajuƒáom granom).
2. Nastavi dok ne ostanu dva ƒçvora ‚Äî oni se ne ukljuƒçuju u kod.

Na kraju se dobija niz od taƒçno $n - 2$ elemenata.

```python
import networkx as nx

def prufer_encode(tree_edges):
    """
    Prima: listu grana stabla (npr. [(1,2), (1,3), (3,4), ...])
    Vraƒáa: Priferov kod kao listu celih brojeva
    """
    import heapq
    from collections import defaultdict

    # Skup svih ƒçvorova
    nodes = set()
    for u, v in tree_edges:
        nodes.add(u)
        nodes.add(v)

    n = len(nodes)
    degree = defaultdict(int)
    graph = defaultdict(set)

    for u, v in tree_edges:
        degree[u] += 1
        degree[v] += 1
        graph[u].add(v)
        graph[v].add(u)

    # Min-heap svih listova (ƒçvorovi sa stepenom 1)
    leaves = [node for node in nodes if degree[node] == 1]
    heapq.heapify(leaves)

    code = []
    for _ in range(n - 2):
        leaf = heapq.heappop(leaves)
        neighbor = next(iter(graph[leaf]))

        code.append(neighbor)

        degree[neighbor] -= 1
        degree[leaf] -= 1

        graph[neighbor].remove(leaf)
        graph[leaf].remove(neighbor)

        if degree[neighbor] == 1:
            heapq.heappush(leaves, neighbor)

    return code

# Primer poziva:
edges = [(1, 2), (1, 3), (2, 4), (2, 5), (2, 6), (6, 7)]
prufer_code = prufer_encode(edges)
print("Priferov kod:", prufer_code)
```

**_Napomena:_**  
Za neorijentisani graf $G=(V,E)$, i ƒçvor $v‚ààV$, broj ponavljanja etikete ƒçvora u Priferovom kodu jednak je $deg(v) - 1$ . U ovom primeru, ƒçvor $2$ se pojavljuje 3 puta jer $deg(2) = 4$ (povezan je sa ƒçvorovima 1, 4, 5, i 6, posredno). Ovo je dodatna provera doslednosti algoritma.

**_Teorema:_**

Broj etiketiranih stabala

Za dati broj $n$ ƒçvorova, broj razliƒçitih etiketiranih stabala jednak je:

$$
n^{n - 2}
$$

Ova formula je poznata kao **Cayleyjeva formula**, a direktna posledica ƒçinjenice da postoji bijekcija izmeƒëu etiketiranih stabala i Priferovih kodova du≈æine $n - 2$ sa elementima iz $\{1, 2, \dots, n\}$.


---

### **Algoritam za rekonstrukcija stabla iz Priferovog koda**

Neka je dat Priferov kod:

$$
P = (a_1, a_2, \dots, a_{n-2})
$$

gde je svaki $a_i \in \{1, 2, \dots, n\}$.

Koraci za rekonstrukciju etiketiranog stabla sa $n$ ƒçvorova su sledeƒái:

1. Inicijalizuj niz $D$ du≈æine $n$, gde je $D[v]$ broj pojavljivanja ƒçvora $v$ u Priferovom kodu $P$.
2. Za sve ƒçvorove $v \in \{1, \dots, n\}$ koji **ne postoje** u $P$, postavi $D[v] = 0$.
3. Napravi pomoƒánu strukturu (npr. prioritetni red) da efikasno ƒçuva ƒçvorove sa $D[v] = 0$ ‚Äî to su trenutno **listovi**.
4. Za svaki element $a_i$ u kodu, redom:
   - Uzmi **najmanji ƒçvor $v$** takav da je $D[v] = 0$ (najmanji dostupni list).
   - Dodaj granu $(v, a_i)$.
   - A≈æuriraj $D[a_i] := D[a_i] - 1$, a $D[v] := -1$ (ili ga ukloni iz razmatranja).
   - Ako nakon umanjenja $D[a_i]$ postane 0, dodaj $a_i$ u skup listova.
5. Kada obradi≈° sve elemente niza, ostaju jo≈° **dva ƒçvora sa $D[v] = 0$** ‚Äî pove≈æi ih poslednjom granom.

Na kraju dobija≈° povezano stablo sa $n$ ƒçvorova i $n - 1$ grana.


_Primer_ :

**Rekonstrukcija stabla iz Priferovog koda**

Neka je dat Priferov kod:

$$
P = (1, 2, 2, 2, 6)
$$

**Korak 1:** Odredi broj ƒçvorova

Priferov kod du≈æine $n - 2 = 5$ implicira da je broj ƒçvorova:

$$
n = 7
$$

Dakle, skup ƒçvorova je:

$$
V = \{1, 2, 3, 4, 5, 6, 7\}
$$

**Korak 2:** Inicijalizuj broj pojavljivanja (frekvencije)

| ƒåvor | Broj pojavljivanja u kodu | Inicijalni stepen |
| ---- | ------------------------- | ----------------- |
| 1    | 1                         | 2                 |
| 2    | 3                         | 4                 |
| 3    | 0                         | 1                 |
| 4    | 0                         | 1                 |
| 5    | 0                         | 1                 |
| 6    | 0                         | 1                 |
| 7    | 1                         | 2                 |

(Stepen = broj pojavljivanja + 1)

**Korak 3:** Iterativno formiraj grane

Inicijalni skup listova (ƒçvorovi sa $0$ pojavljivanja): $\{3, 4, 5, 6\}$  
Sortirano: $[3, 4, 5, 6]$

Obraƒëujemo kod redom: $(1, 2, 2, 2, 6)$

- **Prvi element:** $1$  
  Najmanji list: $3$ ‚áí grana $(3, 1)$  
  A≈æuriramo: $D[3] = -1$, $D[1] = 0$ ‚áí dodajemo $1$ u listove

- **Drugi element:** $2$  
  Najmanji list: $1$ ‚áí grana $(1, 2)$  
  A≈æuriramo: $D[1] = -1$, $D[2] = 2$

- **Treƒái element:** $2$  
  Najmanji list: $4$ ‚áí grana $(4, 2)$  
  A≈æuriramo: $D[4] = -1$, $D[2] = 1$

- **ƒåetvrti element:** $2$  
  Najmanji list: $5$ ‚áí grana $(5, 2)$  
  A≈æuriramo: $D[5] = -1$, $D[2] = 0$ ‚áí dodajemo $2$ u listove

- **Peti element:** $6$  
  Najmanji list: $2$ ‚áí grana $(2, 6)$  
  A≈æuriramo: $D[2] = -1$, $D[6] = 0$ ‚áí dodajemo $6$ u listove

Na kraju ostaju ƒçvorovi sa $D[v] = 0$: ƒçvorovi $6$ i $7$  
‚áí dodajemo poslednju granu $(6, 7)$

**Dobijeno stablo**

Skup grana rekonstruisanog stabla:

$$
\{(3,1), (1,2), (4,2), (5,2), (2,6), (6,7)\}
$$

```python
import networkx as nx
import matplotlib.pyplot as plt
from collections import defaultdict

def prufer_decode(prufer_code):
    """
    Prima: Priferov kod kao listu celih brojeva
    Vraƒáa: listu grana rekonstruisanog stabla
    """
    n = len(prufer_code) + 2
    degree = defaultdict(int)

    for node in prufer_code:
        degree[node] += 1

    # Inicijalni ƒçvorovi
    nodes = set(range(1, n + 1))
    result = []

    for node in prufer_code:
        for leaf in sorted(nodes):
            if degree[leaf] == 0:
                result.append((leaf, node))
                degree[node] -= 1
                nodes.remove(leaf)
                break

    # Preostala dva ƒçvora
    u, v = sorted(nodes)
    result.append((u, v))

    return result

def hierarchy_pos(G, root=None, width=1.0, vert_gap=0.2, vert_loc=0, xcenter=0.5, pos=None, parent=None):
    """
    Hijerarhijski layout za stablo.
    """
    if pos is None:
        pos = {root: (xcenter, vert_loc)}
    else:
        pos[root] = (xcenter, vert_loc)
    children = list(G.neighbors(root))
    if parent is not None:   # ukloni roditelja iz liste dece
        children.remove(parent)
    if len(children) != 0:
        dx = width / len(children)
        next_x = xcenter - width/2 - dx/2
        for child in children:
            next_x += dx
            pos = hierarchy_pos(G, root=child, width=dx, vert_gap=vert_gap,
                                vert_loc=vert_loc - vert_gap, xcenter=next_x, pos=pos, parent=root)
    return pos


def plot_tree(edges):
    """
    Iscrtava stablo zadato listom grana u hijerarhijskom rasporedu.
    """
    import matplotlib.pyplot as plt
    import networkx as nx

    G = nx.Graph()
    G.add_edges_from(edges)

    degree = dict(G.degree())
    root = min([node for node, deg in degree.items() if deg > 1], default=list(G.nodes)[0])

    pos = hierarchy_pos(G, root)

    plt.figure(figsize=(8, 6))
    nx.draw(G, pos, with_labels=True, node_color='lightgreen',
            node_size=1000, font_size=14, edge_color='gray')
    plt.title("Prikaz stabla", fontsize=16)
    plt.show()


# Primer poziva:
code = [1, 2, 2, 2, 6]
edges = prufer_decode(code)
print("Rekonstruisane grane:", edges)
plot_tree(edges)
```

## **Te≈æinski grafovi i minimalna pokrivajuƒáa stabla**

### **Te≈æinski graf**

**_Definicija:_**

> Neka je $G = (V, E)$ konaƒçan, neusmeren graf. Ako je svakoj grani $e \in E$ dodeljena realna vrednost $w(e) \in \mathbb{R}$ (koja predstavlja "tro≈°ak", "du≈æinu", "te≈æinu" ili sliƒçan atribut), tada ka≈æemo da je $G$ **te≈æinski graf**.
> Formalno, imamo funkciju te≈æine:

$$
w : E \rightarrow \mathbb{R}
$$

U kontekstu usmerenih grafova, te≈æine se dodeljuju **lukovima**, tj. parovima $(u, v)$, ≈°to omoguƒáava i razliƒçite vrednosti $w(u, v) \neq w(v, u)$.

### **Minimalno pokrivajuƒáe stablo (MST)**

**_Definicija_**:

> Za povezani te≈æinski neusmeren graf $G = (V, E)$, **minimalno pokrivajuƒáe stablo** (engl. _Minimum Spanning Tree_, skraƒáeno **MST**) je podgraf $T = (V, E')$ takav da:
> - $T$ je stablo: povezan i acikliƒçan,
>- $E' \subseteq E$, i $|E'| = |V| - 1$,
>- Ukupna te≈æina je minimalna: $$\sum_{e \in E'} w(e) = \min$$

MST ne mora biti jedinstven ‚Äî mo≈æe postojati vi≈°e razliƒçitih stabala iste minimalne te≈æine.


## **Algoritmi za odreƒëivanje minimalnog pokrivajuƒáeg stabla**

---

### **1. Kruskalov algoritam**

**Ideja:**

Postupno dodajemo najjeftinije grane koje ne formiraju ciklus, sve dok ne dobijemo stablo sa $n - 1$ grana.

**Koraci:**

1. Sortiraj sve grane rastuƒáe po te≈æini.
2. Inicijalizuj $T = \emptyset$ (nema grana).
3. Za svaku granu $e = (u, v)$ iz sortirane liste:
   - Ako dodavanje $e$ u $T$ **ne stvara ciklus** ‚áí dodaj $e$ u $T$.
4. Zaustavi se kada $|T| = n - 1$.

**Struktura podataka:**

Za detekciju ciklusa koristi se **Union-Find (Disjoint Set Union, DSU)** struktura sa optimizacijama:

- spajanje po rangu
- putna kompresija

![Vizualizacija Kruskalovog algoritma](https://upload.wikimedia.org/wikipedia/commons/5/5c/MST_kruskal_en.gif)



```python
import networkx as nx

def kruskal_mst(G):
    """
    Kruskalov algoritam za pronala≈æenje minimalnog pokrivajuƒáeg stabla (MST).
    Ulaz: neusmeren te≈æinski graf G (networkx.Graph)
    Izlaz: skup grana MST kao lista (u, v, te≈æina)
    """
    # Inicijalizuj disjunktne skupove (Union-Find)
    parent = {}
    def find(u):
        while parent[u] != u:
            parent[u] = parent[parent[u]]
            u = parent[u]
        return u
    def union(u, v):
        ru, rv = find(u), find(v)
        if ru != rv:
            parent[ru] = rv
            return True
        return False

    for node in G.nodes:
        parent[node] = node

    mst_edges = []
    sorted_edges = sorted(G.edges(data=True), key=lambda e: e[2].get('weight', 1))

    for u, v, data in sorted_edges:
        if union(u, v):
            mst_edges.append((u, v, data.get('weight', 1)))
            if len(mst_edges) == G.number_of_nodes() - 1:
                break

    return mst_edges

# Primer poziva:
G = nx.Graph()
G.add_weighted_edges_from([
    (1, 2, 4),
    (1, 3, 5),
    (2, 3, 1),
    (3, 4, 3),
    (2, 4, 2)
])

print("Kruskal MST:", kruskal_mst(G))
```

### **2. Primov algoritam**

**Ideja:**

Postupno izgraƒëujemo stablo tako ≈°to u svakoj iteraciji dodajemo najjeftiniju granu koja povezuje veƒá izabrani deo grafa sa nekim neizabranim ƒçvorom.

**Koraci:**

1. Odaberi proizvoljan poƒçetni ƒçvor $s \in V$.
2. Inicijalizuj $T = \{s\}$, a skup $E_T = \emptyset$.
3. Dok $T \neq V$:
   - Naƒëi granu $(u, v)$ sa najmanjom te≈æinom takvu da je $u \in T$ i $v \notin T$.
   - Dodaj ƒçvor $v$ i granu $(u,v)$ u $T$.

**Struktura podataka:**

Efikasna implementacija koristi **min-heap (prioritetni red)** za odr≈æavanje najjeftinijih kandidata za pro≈°irenje.

![Vizualizacija Primovog algoritma](https://upload.wikimedia.org/wikipedia/commons/9/9b/PrimAlgDemo.gif)


```python
import heapq

def prim_mst(G, start=None):
    """
    Primov algoritam za minimalno pokrivajuƒáe stablo (MST).
    Ulaz: te≈æinski neusmeren graf G (networkx.Graph)
    start: poƒçetni ƒçvor (ako nije dat, koristi najmanji ƒçvor)
    Izlaz: lista grana (u, v, weight)
    """
    if start is None:
        start = list(G.nodes)[0]

    visited = set()
    mst_edges = []
    min_heap = []

    def add_edges(node):
        visited.add(node)
        for neighbor, data in G[node].items():
            if neighbor not in visited:
                weight = data.get('weight', 1)
                heapq.heappush(min_heap, (weight, node, neighbor))

    add_edges(start)

    while min_heap and len(visited) < G.number_of_nodes():
        weight, u, v = heapq.heappop(min_heap)
        if v not in visited:
            mst_edges.append((u, v, weight))
            add_edges(v)

    return mst_edges

# Primer poziva:
G = nx.Graph()
G.add_weighted_edges_from([
    (1, 2, 4),
    (1, 3, 5),
    (2, 3, 1),
    (3, 4, 3),
    (2, 4, 2)
])
print("Prim MST:", prim_mst(G, start=1))

```

## **Najkraƒái putevi u grafu**

**_Definicija:_**

> Za te≈æinski **usmeren** ili **neusmeren** graf $G = (V, E)$ sa funkcijom te≈æine $w : E \to \mathbb{R}_{\geq 0}$, **najkraƒái put** izmeƒëu ƒçvorova $s$ i $v$ je put $P = (s, \dots, v)$ koji minimizuje:

$$
w(P) = \sum_{e \in P} w(e)
$$


### **Dijkstrin algoritam**

**Uslov:**

Sve te≈æine u grafu moraju biti **nenegativne**: $w(e) \geq 0$ za sve $e \in E$.

**Ulaz:**

- Graf $G = (V, E)$
- Izvorni ƒçvor $s \in V$

**Izlaz:**

- Niz $d[v]$ ‚Äî najkraƒáa rastojanja od $s$ do svakog $v \in V$

**Koraci algoritma:**

1. Inicijalizuj rastojanja:
   $$
   d[v] :=
   \begin{cases}
     0, & \text{ako } v = s \\
     \infty, & \text{inaƒçe}
   \end{cases}
   $$
2. Inicijalizuj prioritetni red $Q$ sa svim ƒçvorovima (po $d[v]$).
3. Dok $Q$ nije prazan:
   - Ekstrahuj ƒçvor $u$ sa najmanjom vredno≈°ƒáu $d[u]$ iz $Q$.
   - Za svaku susednu granu $(u, v)$:
     - Ako $d[v] > d[u] + w(u,v)$:
       - A≈æuriraj $d[v] := d[u] + w(u,v)$
       - A≈æuriraj poziciju $v$ u $Q$

**Slo≈æenost:**

- $O((|V| + |E|) \log |V|)$ uz **Fibonacci heap** (ili $O(|E| \log |V|)$ uz **binary heap**)

![Vizualizacija Dijkstrinog algoritma](https://upload.wikimedia.org/wikipedia/commons/5/57/Dijkstra_Animation.gif)



```python
import heapq

def dijkstra_shortest_path(G, start):
    """
    Dijkstrin algoritam za najkraƒáe puteve (nenegativne te≈æine).
    Ulaz: G - te≈æinski graf (networkx), start - poƒçetni ƒçvor
    Izlaz: rastojanja od start do svih drugih ƒçvorova
    """
    dist = {node: float('inf') for node in G.nodes}
    dist[start] = 0

    visited = set()
    heap = [(0, start)]

    while heap:
        cost, u = heapq.heappop(heap)
        if u in visited:
            continue
        visited.add(u)

        for v, data in G[u].items():
            weight = data.get('weight', 1)
            if dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight
                heapq.heappush(heap, (dist[v], v))

    return dist

# Primer:
G = nx.DiGraph()
G.add_weighted_edges_from([
    (1, 2, 6),
    (1, 3, 5),
    (1, 4, 5),
    (2, 5, 1),
    (3, 2, 2),
    (3, 5, 1),
    (4, 3, 2),
    (4, 6, 1),
    (5, 7, 3),
    (6, 7, 3)
])
print(dijkstra_shortest_path(G, 1))
```

### **Bellman-Ford algoritam za najkraƒái put u grafu**

Bellman-Ford algoritam re≈°ava problem najkraƒáeg puta iz jednog izvora $s$ ka svim ostalim ƒçvorovima u te≈æinskom grafu $G = (V, E)$, ƒçak i kada **neke te≈æine mogu biti negativne**.

Za razliku od Dijkstrinog algoritma, koji zahteva da sve te≈æine budu nenegativne, Bellman-Ford funkcioni≈°e i kada:

$$
\exists e \in E \text{ tako da } w(e) < 0
$$

**Ulaz:**

- Usmeren ili neusmeren te≈æinski graf $G = (V, E)$ (mo≈æe imati negativne te≈æine)
- Poƒçetni ƒçvor $s \in V$

**Izlaz:**

- Niz $d[v]$ najkraƒáih rastojanja od $s$ do svakog $v \in V$
- Ako postoji negativan ciklus koji je dosti≈æan iz $s$, algoritam to detektuje

**Koraci algoritma:**

1. Inicijalizuj rastojanja:
   $$
   d[v] :=
   \begin{cases}
   0, & v = s \\
   \infty, & \text{inaƒçe}
   \end{cases}
   $$
2. **Ponovi $|V| - 1$ puta:**
   - Za svaku granu $(u, v)$:
     - Ako $d[u] + w(u, v) < d[v]$:
       - a≈æuriraj $d[v] := d[u] + w(u, v)$
3. **Detekcija negativnog ciklusa:**
   - Ako i nakon $|V| - 1$ iteracija postoji grana $(u, v)$ koja mo≈æe a≈æurirati $d[v]$, tada postoji **negativni ciklus**

**Slo≈æenost:**

- Vremenska: $O(|V||E|)$
- Prostorna: $O(|V|)$

**Kori≈°ƒáenje:**

- Kada graf mo≈æe imati negativne te≈æine
- Kada je potrebno detektovati negativne cikluse
- U mre≈ænim algoritmima, dinamiƒçkom programiranju, teoriji valova itd.

![Vizualizacija Bellman-Ford algoritma](https://upload.wikimedia.org/wikipedia/commons/7/77/Bellman%E2%80%93Ford_algorithm_example.gif)



```python
def bellman_ford_shortest_path(G, start):
    """
    Bellman-Ford algoritam za najkraƒáe puteve (dozvoljene negativne te≈æine).
    Ulaz: G - te≈æinski usmeren graf (networkx.DiGraph), start - poƒçetni ƒçvor
    Izlaz: dict rastojanja, ili Exception ako postoji negativan ciklus
    """
    dist = {node: float('inf') for node in G.nodes}
    dist[start] = 0

    # |V| - 1 iteracija
    for _ in range(len(G.nodes) - 1):
        for u, v, data in G.edges(data=True):
            weight = data.get('weight', 1)
            if dist[u] + weight < dist[v]:
                dist[v] = dist[u] + weight

    # Detekcija negativnog ciklusa
    for u, v, data in G.edges(data=True):
        weight = data.get('weight', 1)
        if dist[u] + weight < dist[v]:
            raise ValueError("Graf sadr≈æi negativan ciklus.")

    return dist

# Primer:
G = nx.DiGraph()
G.add_weighted_edges_from([
    (1, 2, 6),
    (1, 3, 5),
    (1, 4, 5),
    (2, 5, -1),
    (3, 2, -2),
    (3, 5, 1),
    (4, 3, -2),
    (4, 6, -1),
    (5, 7, 3),
    (6, 7, 3)
])
print(bellman_ford_shortest_path(G, 1))
```
