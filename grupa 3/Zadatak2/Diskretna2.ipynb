# **1.	Klasifikacija uređenih izbora elemenata**

-	Uređeni izbori elemenata se mogu klasifikovati na sleddeće načine.

**Uređenje m elemenata iz skupa od n elemenata:**

*   **m-permutacije skupa**. $(m \le n)$. Bilo koja $m-torka$ početnog skupa gde su svaka dva elementa međusobno različita.
*   **Permutacije skupa**. Specijalan slučaj m-permutacija skupa kada $(m = n)$.

**Uređenje m elemenata iz multiskupa koji ima n elemenata:**


*   **m-permutacije multiskupa** $(m \le n, M = [b_1, …. b_l]m_1, ….., m_l )$. Bilo koja uređena $m-torka$ gde je svaka komponenta iz skupa $M$.
*   **permutacije multiskupa** $(m = n, M = [b_1, ….., b_l]m_1, …., m_l)$.  Proizvoljna $n-torka$ u kojoj se $b_1$ pojavljuje $m_1$ puta, $b_2$ $m_2$ puta, …, $b_1$ pojavljuje $m_l$ puta.

# **2. Varijacije sa ponavljanjem (m-permutacija multiskupa)**

* Neka je dat skup $B = \{b_1, ...., b_l\}$ sa $l \ge 1$ elemenata i neka je $M = [b_1, ..., b_l]m_1, ..., m_l$ multiskup u kojem se se svaki element iz $B$ pojavljuje tačno $m$ puta. Broj elemenata u multiskupu $M$ je $ n = m * l$. Ako izaberemo $m-točlani$ podmultiskup od $M$ i uredimo ga da dobijamo jednu m-permutaciju multiskupa.
* **Definicija 12**
$m-permutacija$ elementa multiskupa $M = [b_1, ..., b_l]m_1, ..., m_l$ je bilo koja uređena $m-torka$ elemenata iz $M$, tj. bilo koja uređena $m-torka$ u kojoj je svaka komponenta iz skupa $B$.
* **Teorema 13** Broj m-permutacija elemenata multiskupa $M = [b_1, ..., b_l]m_1, ..., m_l$ elemenata je $l^m$,
  * $\overline{P}(l; m) = l^m$  .



*Dokaz*. Prema definiciji 12, svaki element skupa $B \times ... \times B $ predstavljajednu m-permutaciju multiskupa $M$. Broj takvih $m-torki$ elemenata iz $B$ je na osnovu principa proizvoda:

   *	$| B \times  .... \times B| = |B|^m = l^m$ .

# **3. Varijacije bez ponavljanja (m-permutacija skupa)**

* Neka je dat skup $B = \{b_1, ..., b_n\}$ i neka je 1 $\le$ $m$ $\le$ $n$ .

* **Definicija 15** $m-permutacija$ elemenata skupa $B$ je bilo koja $m-torka$ elemenata skupa $B$ u kojo su svaka dva elementa međusobno različita.

* **Teorema 16** Broj $m-permutacija$ skupa $B = \{b_1, ..., b_n\}$  jednak je  $P(n; m) = n(n - 1) ... (n - m + 1).$

# Permutacije

## Permutacije skupa

U kombinatorici, **permutacije** skupa od $n$ elemenata predstavljaju sve moguće nizove dužine $n$ koji se mogu dobiti ređanjem elemenata posmatranog skupa gde se svaki pojavljuje tačno jednom.

Očigledno je da su ovako definisane permutacije samo specijalan slučaj $m$-permutacija skupa $n$ elemenata i da postoji analogija u načinu prebrojavanja.

Koristeći formulu za prebrojavanje varijacija ($m$-permutacija skupa) $P(n;m) = \frac{n!}{(n-m)!}$ i ubacivanjem $m = n$ dobijamo $P(n) = \frac{n!}{(n-n)!} = n!$.

**Primer**

Napisati python funkciju koja nabraja sve permutacije skupa.

```python
def permute(nums):
        ans = []
        if len(nums) == 1:
            return [nums[:]]
        for num in nums:
            tmp = nums.pop(0)
            perms = permute(nums)
            for perm in perms:
                perm.append(tmp)
            ans += perms
            nums.append(tmp)
        return ans

perms = permute([1,2,3,4])
print(perms)
print("Broj permutacija: ", len(perms))
```

Svaka permutacija predstavlja jedinstvenu bijekciju koja slika skup u samog sebe. Prema tome, broj permutacija skupa $A$ predstavlja i broj različitih bijektivnih funkcija $f: A→A$. Takođe, iz činjenice da kompozicija dve permutacije dobijamo novu sledi da je kompozicija dve bijekcije isto bijekcija.

## Permutacije multiskupa

U slučaju multiskupa, permutacije se definišu na isti način. Imajući u vidu da multiskup može sadržati više istih elemenata, formula kojom dobijamo broj permutacija multiskupa je nešto drugačija.

Neka $m_1,m_2,...,m_l$ predstavljaju učestalost svakog od $l$ različitih elemenata multiskupa $M$. Tada se broj permutacija ovog multiskupa računa kao:

$P(m_1,m_2,...,m_l) = \frac{(m_1+m_2+…+m_l)!}{m_1!m_2! … m_l!}$.

Bitno je primetiti da zbir u brojiocu predstavlja ukupan broj elemenata multiskupa. S obzirom na to, može se uvideti sličnost permutacija skupa i multiskupa.

**Primer**

Prava prizma dimenzija $3\times4\times6$ je izdeljena na kockice stranice $1cm$. Ukoliko se možemo kretati po svakoj od tri ose i nalazimo se u kockici koja sadrži teme prizme, na koliko načina je moguće doći do suprotnog temena.

**Rešenje**

Pod pretpostavkom da ćemo se uvek kretati ka našem cilju (suprotnom temenu) biće nam potrebno da napravimo ukupno $3$ koraka po $x$-osi, $4$ po $y$ i $6$ po $z$-osi. Redosled kojim to činimo je bitan jer on određuje različite puteve.

Ovaj problem se svodi na permutacije multiskupa $\{x,x,x,y,y,y,y,z,z,z,z,z,z\}$. Dakle broj mogućih puteva ka suprotnom temenu možemo izračunati kao $\frac{(3+4+6)!}{3!4!6!} = \frac{13!}{3!4!6!}$.
