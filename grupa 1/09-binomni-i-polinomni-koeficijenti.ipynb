# Binomni i polinomni koeficijenti i formule

## Binomni koeficijent

Binomni koeficijenti predstavljaju koeficijente u razvoju stepena binoma u zbir prema binomnoj teoremi.

### _Definicija:_ 
Neka su $m$ i $n$ celi brojevi sa osobinom $0 \leq m \leq n$. Binomni koeficijent $\binom{n}{m}$ je funkcija koja takvim parovima vrednosti $n$ i $m$ dodeljuje pozitivne cele brojeve na sledeći način:

$$\binom{n}{m} = \frac{n\left(n - 1\right ) ... \left(n-m+1\right)}{m\left(m-1\right)...\ 2\cdot 1}, \ m \geq 1$$ 

Binomni koeficijent se kombinatorno može interpretirati kao broj $m$-kombinacija skupa od $n$ elemenata, odnosno kao broj $m$-točlanih podskupova skupa od $n$ elemenata.

Kombinatorna interpretacija formalno se može opisati relacijom 
$$\dbinom{n}{m} = C\left(n; m\right)$$

### _Faktorijel reprezentacija:_
Za cele brojeve $n$ i $m$ sa osobinom $0 \leq m \leq n$, važi:
$$\dbinom{n}{m} = \frac{n!}{m!\left(n-m\right)!}$$
*$Dokaz:$* Za $m \in \left\{0, n\right\}$ imamo:
$$\dbinom{n}{n} = \frac{n!}{n!} = 1 = \dbinom{n}{0}$$
Ako je $1 \leq m \leq n-1$ množenjem brojioca i imenioca sa $\left(n-m\right)!$ dobijamo:
$$\dbinom{n}{m} = \frac{n\left(n-1\right)...\left(n-m+1\right)\left(n-m\right)!}{m\left(m-1\right)...\ 2\cdot 1\cdot \left(n-m\right)!} = \frac{n!}{m!\left(n-m\right)!}$$


<br/>

$$\tag*{□}$$
<br/>

Možemo primetiti da binomni koeficijent $\binom{n}{m}$ odgovara broju kombinacija bez ponavljanja klase $m$ od $n$ elemenata. Broj načina da se od $n$ elemenata izabere $m$ elemenata jednak je broju načina da se od $n$ elemenata izabere preostalih $n-m$ elemenata. To je pokazano u osobini simetričnosti.


```python
def bin_koef(n: int, m: int) -> int:
    top = 1
    bottom = 1
    for i in range(1, m + 1):
        top *= n - m + i
        bottom *= i
    return top // bottom
```

```python
print(bin_koef(5, 3))
```

### Osobine binomnih koeficijenata

$$\dbinom{n}{0} = 1$$

_Alegbarski:_

$$
    \dbinom{n}{0} = \frac{n!}{0!(n-0)!} = \frac{\cancel{n!}}{1\cancel{n!}} = 1
$$

_Kombinatorno:_

Na samo jedan način možemo odabrati 0 elemenata od mogućih $n$.

$$\dbinom{n}{1} = n$$

_Alegbarski:_

$$
    \dbinom{n}{1} = \frac{n!}{1!(n-1)!} = \frac{n\cancel{(n-1)!}}{1\cancel{(n-1)!}} = n
$$

_Kombinatorno:_

Od ponuđenih $n$ elemenata možemo izabrati jedan element na $n$ načina, odnosno svaki element pojedinačno.

$$\dbinom{n}{n} = 1$$

_Alegbarski:_

$$
    \dbinom{n}{n} = \frac{n!}{n!(n-n)!} = \frac{\cancel{n!}}{\cancel{n!}1} = 1
$$

_Kombinatorno:_

Od ponuđenih $n$ elemenata možemo izabrati $n$ element na 1 način tako što odaberemo sve elemente.



#### Simetričnost
Ako je $n$ prirodan broj i $k \in \Z$ tada je
$${n \choose k} = {n \choose n-k}$$
gde je $n \choose k$ binomni koeficijent

_Dokaz:_

Ukoliko je $k < 0$ tada je i $n−k > n$  
Ukoliko je $k > n$ tada je i $n−k < 0$
U oba navedena slučaja važi
$${n \choose k} = {n \choose n-k} = 0$$

Neka $k$ bude takvo da $0 ≤ k ≤ n$. Tada važi
$${n \choose k} = {n! \over {k! (n-k)!}}$$
$$= {n! \over {(n-k)!k!}}$$
$$= {n! \over {(n-k)!(n-(n-k))!}}$$
$$= {n \choose {n-k}}$$

<br/>

$$\tag*{□}$$

<br/>

#### Kombinatorna interpretacija
Binomni koeficijent u kombinatorici ima primenu da odredi na koliko načina se može izabrati $k$ elemenata iz skupa od njih $n$

#### Delioci binomnog koeficijenta
za svako $r \in \R, k \in \Z$ tada važi
$$k {r \choose k} = r {r-1 \choose k-1}$$
gde $r \choose k$ predstavlja binomni koeficijent

Odatle sledi
$${r \choose k} = {r \over k}{r-1 \choose k-1}$$
odnosno
$${1 \over r}{r \choose k} = {1 \over k}{r-1 \choose k-1}$$

_Dokaz:_

Ukoliko je k=0 tada je po definiciji
$${r \choose k} = r{r-1 \choose k-1} = 0$$

Kada $k \ne 0$ tada je
$$k {r \choose k} = k{r! \over (r-k)!k!}$$
$$= k{r(r−1)(r−2)⋯(r−k+1) \over k(k−1)(k−2)⋯1}$$
$$= {r(r−1)(r−2)⋯(r−k+1) \over (k−1)(k−2)⋯1}$$
$$= r{r-1 \choose k-1}$$
Kako $k \ne 0$ možemo podeliti obe strane sa $k$ i dobijamo
$${r \choose k} = {r \over k}{r-1 \choose k-1}$$
Ukoliko je i $r \ne 0$ tada važi i
$${1 \over r}{r \choose k} = {1 \over k}{r-1 \choose k-1}$$


<br/>

$$\tag*{□}$$

<br/>


#### Paskalov identitet
Neka je $n \choose k$ binomni koeficijent. Tada za pozitivne cele brojeve $n, k$ takve da $1 \leq k \leq n$
$${n \choose k-1} + {n \choose k} = {n+1 \choose k}$$

_Dokaz:_

Neka su $n, k \in \N$ takvi da $1 \leq k \leq n$
$${n \choose k} + {n \choose k-1} = {n! \over {k!(n-k)!}} + {n! \over {k-1}!(n-(k-1))!}$$
$$= {n!(n-(k-1)) \over {k!(n-k)!(n-(k-1))}} + {n! k \over {k-1}!(n-(k-1))!k}$$
$$= {n!(n-k+1) \over {k!(n-k+1)!}} + {n! k \over k!(n-k+1)!}$$
$$= {n!(n-k+1) + n! k \over {k!(n-k+1)!}}$$
$$= {n!(n-k+1+k) \over {k!(n-k+1)!}}$$
$$= {n!(n+1) \over {k!(n-k+1)!}}$$
$$= {(n+1)! \over {k!(n+1-k)!}}$$
$$= {n+1 \choose k}$$


<br/>

$$\tag*{□}$$

<br/>


_Kombinatorni dokaz_

Recimo da postoji klub od $n+1$ članova koji treba da izabere upravni odbor od $k$ članova. Prema kardinalitetu skupa svih podskupova načina da se odabere odbor postoji tačno
$$n+1 \choose k$$

_Koristimo drugi pristup (Takođe koristimo kardinalitet skupa svih podskupova):_

Primetimo neku osobu $A$ u tom klubu. Ukoliko je osoba $A$ izabrana za upravni odbor, tada biramo jos $k-1$ osoba od njih n, odnosno imamo $n \choose k-1$ mogućnosti
Ukoliko osoba $A$ nije izabrana za upravni odbor, tada biramo svih $k$ članova od preostalih $n$ i to možemo na $n \choose k$ načina.

Ova dva slučaja su disjunktna te principom zbira dobijamo da je ukupan broj načina na koji se može izabrati upravni odbor
$${n \choose k} + {n \choose k-1}$$

Prvi i drugi način moraju dati ista rešenja, prema tome
$${n+1 \choose k} = {n \choose k} + {n \choose k-1}$$

<br/>

$$\tag*{□}$$

<br/>

#### Vandermondov identitet

$$\binom{m}{0}\binom{n}{k} + \binom{m}{1}\binom{n}{k-1} + ... + \binom{m}{k}\binom{n}{0} = \binom{m+n}{k}$$

_Kombinatorni dokaz:_

Ukoliko imamo $m$ matematičara i $n$ inženjera od kojih želimo da oformimo tim od $k$ osoba.

To možemo uraditi tako što ćemo odabrati 0 matematičara na $\binom{m}{0}$ načina i $k$ inženjera na $\binom{n}{k}$ načina, ili 1 matematičara na $\binom{m}{1}$ načina i $k-1$ inženjera na $\binom{n}{k-1}$ načina ili ... ili $k$ matematičara na $\binom{m}{k}$ i 0 inženjera na $\binom{k}{0}$ načina.
S obzirom da su ovo disjunktni slučajevi (imamo ili izmenju) onda je ukupan broj načina da se odabere tim

$$\binom{m}{0}\binom{n}{k} + \binom{m}{1}\binom{n}{k-1} + ... + \binom{m}{k}\binom{n}{0}$$

Drugi način da oformimo ovaj tim jeste da od svih osoba ($m+n$ ljudi) odaberemo $k$ osoba na $\binom{m+n}{k}$ načina.

S obzirom da su ova dva načina ekvivalentna onda mora da važi i

$$\binom{m}{0}\binom{n}{k} + \binom{m}{1}\binom{n}{k-1} + ... + \binom{m}{k}\binom{n}{0} = \binom{m+n}{k}$$

Što je i trebalo dokazati.
<br/>

$$\tag*{□}$$

## Polinomni koeficijent

### _Definicija_
Neka su dati brojevi $m_{1},\, ... \,,m_{l} \in \N_{0}$ i neka je $n = m_{1}+...+m_{l}$. Tada polinomni koeficijent definišemo na sledeći način:
$$\dbinom{n}{m_{1}, m_{2},\, ... \,,m_{l}} = \frac{n!}{m_{1}!\cdot ...\cdot m_{l}!}$$
Polinomni koeficijent kombinatorno se može interpretirati kao: 
- broj permutacija multiskupa $M = \left[a_{1},...,a_{l}\right]_{m_{1},...,m_{l}}$
- broj uređenih $l$-torki $\left(B_{1},...,B_{l}\right)$ skupa $A = \left\{a_{1},...,a_{l}\right\}$ sa osobinom da je za date vrednosti $\left(m_{1},...,m_{l}\right):$
    - $A = B_{1} \cup ... \cup B_{l}$,
    - $\left|B_{1}\right| = m1,\,...\,,\left|B_{l}\right| = m_{l}$,
    - $m_{1}+...+m_{l} = n$.

Na osnovu prethodne definicije, direktno slede sledeće osobine
### Osobine polinomnog koeficijenta

Neka su dati celi brojevi $m_1, m_2, ..., m_l \ge 0$ i neka je $n = m_1+m_2+...+m_l$

#### _Teorema 1_

$$\dbinom{n}{m_1, m_2, ..., m_l} = \dbinom{n}{m_1}\dbinom{n-m_1}{m_2}\dbinom{n-(m_1+m_2)}{m_3}...\dbinom{m_l}{m_l}$$

_Alegbarski dokaz:_

Ako se primeni definicija binomnog koeficijenta na desnu stranu, po jedan činilac iz imenioca se uvek skrati sa brojiocem narednog razlomka:

$$
\begin{aligned}
\dbinom{n}{m_1}\dbinom{n-m_1}{m_2}...\dbinom{m_l}{m_l} &= 
\frac{n!}{m_1!\cancel{(n-m_1)!}}\frac{\cancel{(n-m_1)!}}{m_2!\cancel{(n-m_1-m_2)!}}...\frac{\cancel{m_l!}}{m_l!0!} \\
&= \frac{n!}{m_1!m_2!...m_l!} \\
&= \dbinom{n}{m_1, m_2, ..., m_l}
\end{aligned}
$$

$$\tag*{□}$$

<br/>

_Kombinatorni dokaz:_

Kombinatorni dokaz se zasniva na postupnom izboru elemenata, grupu po grupu. Pogledajmo kako ovaj proces funkcioniše korak po korak:
1.	Prvi korak: Imamo $n$ elemenata i želimo da odaberemo $m_1$ elemenata za prvu grupu. Broj načina da se izabere $m_1$ elemenata iz skupa od $n$ elemenata je: 
$$\dbinom{n}{m_1}$$

Dakle, prvo biramo $m_1$ elemenata za prvu grupu.
2.	Drugi korak: Nakon što smo odabrali $m_1$ elemenata, ostaje nam $n−m_1$ elemenata. Sada biramo $m_2$ elemenata za drugu grupu iz preostalih $n−m_1$ elemenata. Broj načina da to uradimo je: 
$$\dbinom{n-m_1}{m_2}$$
3.	Treći korak: Nakon što smo odabrali $m_1$ za prvu grupu i $m_2$ za drugu grupu, preostaje $n−(m_1+m_2)$ elemenata. Sada biramo $m_3$ elemenata za treću grupu. Broj načina za ovo je: 
$$\dbinom{n-m_1-m_2}{m_3}$$
4.	Nastavljamo ovaj proces dok ne rasporedimo sve preostale elemente u grupe. Na kraju, kada raspoređujemo poslednjih $m_l$ elemenata u poslednju grupu, imamo preostalo tačno $m_l$ elemenata, pa za taj izbor postoji samo jedan način.

_Skupljanje izbora_

Svaki korak u ovom postupku je nezavisan, što znači da možemo množiti broj načina za svaku fazu izbora:
$$
\dbinom{n}{m_1}\dbinom{n-m_1}{m_2}\dbinom{n-(m_1+m_2)}{m_3}...\dbinom{n-(m_1+m_2+...+m_{l-1})}{m_l}
$$

Ova formula zapravo daje ukupan broj načina da se $n$ elemenata rasporedi u $l$ grupa veličine $m_1,m_2,…,m_l$.

_Zbog čega ovo važi?_

Posmatrajmo sada kompletnu raspodelu:
•	Prvo biramo $m_1$ elemenata iz $n$,
•	Zatim biramo $m_2$ iz preostalih $n−m_1$,
•	Zatim $m_3$ iz $n−(m_1+m_2)$, i tako dalje.

Kada se sve to uzme u obzir, ukupni broj nacina da se svi ovi izbori izvrse za sve $l$ grupa jednak je ukupnom polinomnom koeficijentu $\dbinom{n}{m_1,m_2, ..., m_l}$,koji predstavlja broj nacina da se izvrsi kompletna raspodela odjednom.

Dakle, dokazali smo da je:

$$\dbinom{n}{m_1, m_2, ..., m_l} = \dbinom{n}{m_1}\dbinom{n-m_1}{m_2}\dbinom{n-(m_1+m_2)}{m_3}...\dbinom{m_l}{m_l}$$

$$\tag*{□}$$

<br/>

#### _Teorema 2_

Ako je $\{\{m_1, m_2, ..., m_l\}\} = \{\{k_1, k_2, ..., k_l\}\}$, onda je
$$\dbinom{n}{m_1, m_2, ..., m_l} = \dbinom{n}{k_1, k_2, ..., k_l} $$

_Algebarski dokaz:_

Treba primetit da osobina znaci da se u imeniocu nalaze isti faktorijeli, samo da su mozda uredjeni na razlicite nacine.Kako je mnozenje komutativno I asocijativno, kojim god redom da mnozimo te faktorijele, na kraju cemo dobiti istu vrednost.

$$\tag*{□}$$

<br/>

_Kombinatorni dokaz:_

Jednakost sugeriše da, ako su dva multiskupa (ili podele) $\{\{m_1, m_2, ..., m_l\}\}$ i $\{\{k_1, k_2, ..., k_l\}\}$ jednaka, tada je broj načina da se $n$ objekata raspodeli u grupe veličina $m_1,…,m_l$ isti kao i broj načina da se raspodele u grupe veličina $k_1,…,k_l$.

Jednakost multiskupova: ako su multiksupoviskupovi \{\{m_1, m_2, ..., m_l\}\}$ i $\{\{k_1, k_2, ..., k_l\}\}$ jednaki, to znači da sadrže iste elemente, ali možda u različitom redosledu. Odnosno, implicira da za svaki $i$ postoji $j$ takav da je $m_i=k_j$.
Posto su dva multiskupa identična, redosled u kojem su veličina grupa navedena nije važan. Polinomni koeficijent zavisi samo od veličine grupa, a ne od njihovog rasporeda u formuli. Faktorijali u imenilacu će biti isti (pošto su elementi dva multiskupa isti, iako je drugačiji redosled). Dakle: 
$$\dbinom{n}{m_1, m_2, ..., m_l} = \dbinom{n}{k_1, k_2, ..., k_l}$$


$$\tag*{□}$$

<br/>

#### _Teorema 3_

Ako je $0\lt m_1, m_2, ..., m_l\lt n$ onda važi
$$
\dbinom{n}{m_1, m_2, ..., m_l}  = \dbinom{n-1}{m_1-1, m_2, ..., m_l}  + \dbinom{n-1}{m_1, m_2-1, ..., m_l}  + ... + \dbinom{n-1}{m_1, m_2, ..., m_l-1} 
$$

_Algebarski dokaz:_
$$
\dbinom{n-1}{m_1-1, m_2, ..., m_l}  + \dbinom{n-1}{m_1, m_2-1, ..., m_l}  + ... + \dbinom{n-1}{m_1, m_2, ..., m_l-1} \\
= \frac{(n-1)!}{(m_1-1)!m_2!...m_l!}\frac{m_1}{m_1} + \frac{(n-1)!}{m_1!(m_2 - 1)!...m_l!}\frac{m_2}{m_2} + ... + \frac{(n-1)!}{m_1!m_2!...(m_l-1)!}\frac{m_l}{m_l} \\
= \frac{(n-1)!m_1}{m_1!m_2!...m_l!} + \frac{(n-1)!m_2}{m_1!m_2!...m_l!} + ... + \frac{(n-1)!m_l}{m_1!m_2!...m_l!} \\
= \frac{(n-1)!(m_1 + m_2 + ... + m_l)}{m_1!m_2!...m_l!} \\
= \frac{(n-1)!n}{m_1!m_2!...m_l!} \\
= \frac{n!}{m_1!m_2!...m_l!} \\
= \dbinom{n}{m_1, m_2, ..., m_l}
$$

$$\tag*{□}$$

<br/>

_Kombinatorni dokaz:_

Leva strana odgovara permutacujama sa ponavljanjem $P(m_1,…,m_l)$

Što je broj načina da se uredi $m_1 +…+m_l$ elemenata među kojima se npr. $a_1$ ponavlja $m_1$ puta, $a_2$ ponavlja $m_2$ puta, itd. 

Ako posmatramo desnu stranu, kombinatorna interpretacija je sledeća. Skup svih uređenja mozemo podeliti na $l$ podskupova, tako da svaki podskup sadrži $n$-torke sa istom prvom komponentom. Kako su ti podskupovi po parovima disjunktni, mozemo primeniti princip zbira. Dalje samo treba zakljuciti da je broj načina da se urede preostalih $n-1$ elemenata, pri čemu će biti jedan manje $a_1$ na raspolaganju, a to je $P(m_1-1,m_2,…,m_l)$

Slično se rezonuje o ostalim elementima koji se mogu pojaviti na prvom mestu. 

$$\tag*{□}$$

<br/>




#### _Teorema 4_

$$
\dbinom{n}{m_1, m_2, ..., m_l, 0} = \dbinom{n}{m_1, m_2, ...m m_l}
$$

_Alegbarski dokaz:_

$$
\begin{aligned}
\dbinom{n}{m_1, m_2, ..., m_l, 0} &= \frac{n!}{m_1!m_2!...m_l!0!} \\
&= \frac{n!}{m_1!m_2!...m_l!1} \\
&= \frac{n!}{m_1!m_2!...m_l!} \\
&= \dbinom{n}{m_1, m_2, ...m m_l}
\end{aligned}
$$

$$\tag*{□}$$

<br/>

_Kombinatorni dokaz:_

Desna strana odgovara pemutacijama sa ponavljanjem $P(m_1,…,m_l)$, što je broj načina da se uredi $m_1+…+m_l$ elmenata među kojima se npr. $a_1$ ponavlja $m_1$ puta, $a_2$ ponavlja $m_2$ puta, itd. 

Ako posmatramo levu stranu, kombinatorna interpretacija je sledeća. Skup svih uređenja možemo podeliti na $l+1$ podksupova, tako da svaki podskup sadrzi $n$-torke sa istom prvom komponentom, gde je poslednji $l+1$ podskup zapravo prazan podskup.  Dalje  ostaje isto odgovara pemutacijama sa ponavljanjem  $P(m_1,…,m_l,0)$, što je broj načina da se uredi $m_1+…+m_l$ elmenata medju kojima se npr. $a_1$ ponavlja $m_1$ puta, $a_2$ ponavlja $m_2$ puta, itd a $0!=1$.
$$\tag*{□}$$

## Binomna formula

### _Teorema_

Za sve prirodne brojeve $n$ i $k$, $k\le n$ važi formula:
$$(a+b)^n = \dbinom{n}{0}a^n+\dbinom{n}{1}a^{n-1}b+\dbinom{n}{2}a^{n-2}b^2+...+\dbinom{n}{n-1}a b^{n-1}+\dbinom{n}{n}b^n$$

Kraći zapis bi bio
$$(a+b)^n = \sum\limits_{k=0}^{n}\dbinom{n}{k}a^{n-k}{b^k}$$

Opšti član razvoja bionoma $(a+b)^n$ dat je sa:
$$T_{k+1} = \dbinom{n}{k}a^{n-k}b^{k}\text{ }\text{ za } 0\le k\le n$$

_Dokaz:_

Dokaz ćemo dati indukcijom po $n$

_Baza indukcije:_ Proveravamo da li tvrđenje važi za $n = 0$
$$(a+b)^0 = 1, \sum\limits_{k=0}^{0}\dbinom{0}{k} = \dbinom{0}{0}a^0b^0=1$$

_Induktivna hipoteza:_ Pretpostavimo da trvđenje važi za $n\in\mathbb{N}$ tj. 
$$(a+b)^n = \sum\limits_{k = 0}^{n}\dbinom{n}{k}a^{n-k}b^k$$

_Induktivni korak:_ Dokazaćemo da tvrđenje važi za $n+1$ tj. 
$$(a+b)^{n+1} = \sum\limits_{k = 0}^{n+1}\dbinom{n+1}{k}a^{n+1-k}b^k$$

$$
\begin{aligned}
(a+b)^n+1 &= (a+b)(a+b)^n \\
&= (a+b)(a+b)^n \\
&\overset{\text{BI}}{=} (a+b)\sum\limits_{k=0}^{n}\dbinom{n}{k}a^{n-k}b^k \\
&=a\sum\limits_{k=0}^{n}\dbinom{n}{k}a^{n-k}b^k + b\sum\limits_{k=0}^{n}\dbinom{n}{k}a^{n-k}b^k \\
&=\sum\limits_{k=0}^{n}\dbinom{n}{k}a^{n+1-k}b^k + \sum\limits_{k=0}^{n}\dbinom{n}{k}a^{n-k}b^{k+1} \\
&=\dbinom{n}{0}a^{n+1-0} + \sum\limits_{k=1}^{n}\dbinom{n}{k}a^{n+1-k}b^k + \sum\limits_{k=0}^{n-1}\dbinom{n}{k}a^{n-k}b^{k+1} + \dbinom{n}{n}a^{n-n}b^{n+1} \\
&\overset{\text{k-1 = t}}{=}a^{n+1} + \sum\limits_{k=1}^{n}\dbinom{n}{k}a^{n+1-k}b^k + \sum\limits_{t=1}^{n}\dbinom{n}{t-1}a^{n-t+1}b^{t} + b^{n+1} \\
&\overset{\text{spajamo sume}}{=} a^{n+1} + \sum\limits_{k=1}^{n}(\dbinom{n}{k}+\dbinom{n}{k-1})a^{n+1-k}b^k + b^{n+1} \\
&\overset{\text{Paskal}}{=} a^{n+1} + \sum\limits_{k=1}^{n}\dbinom{n+1}{k}a^{n+1-k}b^k + b^{n+1} \\
&=\sum\limits_{k=0}^{n+1}\dbinom{n+1}{k}a^{n+1-k}b^k
\end{aligned}
$$

## Polinomna formula

### _Teorema_:

Neka su: $ n, k \in \N $. Tada za sve $(x_1,x_2,\dots,x_k\in\mathbb{C}) $ važi:
$$(x_1+x_2+\dots+x_k)^n=\sum_{(n_1,n_2,\dots,n_k)}\binom{n}{n_1\,n_2\dots n_k}x_1^{n_1}x_2^{n_2}\dots x_k^{n_k},$$
pri čemu je svaka $k$-torka $(n_1,n_2,\dots,n_k)$ nenegativnih celih brojeva, takvih da je $n_1+n_2+\dots+n_k=n.$  

_Dokaz:_  

Teoremu ćemo dokazati uopštavanjem kombinatornog dokaza binomne teoreme.
Prvo ćemo izvršiti sva množenja
$$(x_1+x_2+\dots+x_k)^n=(x_1+x_2+\dots+x_k)(x_1+x_2+\dots+x_k)\dots(x_1+x_2+\dots+x_k)$$
Oslobađanjem od zagrada, posle izvršenih $k$ množenja,dobijamo sabirke oblika $ x_1^{n_1}x_2^{n_2}\dots x_k^{n_k} $. Za svaku $k$ -torku $(n_1,n_2,\dots,n_k)$ ne negativnih celih brojeva, takvih da je $n_1+n_2+\dots+n_k=n$ ,dobijamo monom $ x_1^{n_1}x_2^{n_2}\dots x_k^{n_k} $ koji se u dobijenoj sumi pojavljuje kao sabirak onoliko puta koliko imamo mogućnosti da se od ponuđenih $n$ zagrada za sabirak $x_1$ odlučimo $n_1$ puta, za sabira  $x_2$ ,od preostalih $n-n_1$ zagrada, se odlučimo $n_2$ ...  
Prema principu proizvoda zaključujemo da se monom $x_1^{n_1}x_2^{n_2}\dots x_k^{n_k}$ pojavljuje:
$$\binom{n}{n_1}\binom{n-n_1}{n_2}\binom{n-n_1-n_2}{n_3}\dots\binom{n-n_1-n_2-\dots-n_{k-1}}{n_k}=\frac{n!}{n_1!n_2!\dots n_k!}=\binom{n}{n_1\,n_2\dots n_k}\text{ puta}.$$

$$\tag*{□}$$

#### Primer 1

Odrediti koeficijent uz $x^{10}$ u razvoju polinoma $(1 - x^2 + x^3)^{20}$  

_Rešenje:_

Broj 10 se može napisati kao: 
$10 = 5 * 2 = 2 * 2 + 2* 3$  
Sada je potrebno da se izračunaju koeficijenti uz te dve kombinacije:  
U slučeju $ 5 * 2 $ imamo: $ -\binom{20}{15\,5\,0}  $   
  
U drugom slučaju imamo: $\binom{20}{16\,2\,2}$   
  
Sada sabiranjem ovih vrednosti dobijamo rezultat pa bi koeficijent uz $x^{10}$ bio jednak:  

$$ \binom{20}{16\,2\,2} - \binom{20}{15\,5\,0} $$

```python
import sympy as sp
from sympy.parsing.sympy_parser import parse_expr

def expand_polynomial(poly_str, n):
    poly_expr = parse_expr(poly_str)
    expanded_poly = sp.expand((poly_expr)**n)
    
    return expanded_poly

# Example usage
poly_str = "1 - x**2 + x**3"
n = 3
expanded_poly = expand_polynomial(poly_str, n)
print(expanded_poly)
```

## Broj sirjektivnih funkcija skupa A u skup B

### _Teorema:_

Neka su $A$ i $B$ skupovi sa osobinom $|A| = n, |B| = k$ i $1\le k\le n$. Broj sirjektivnih preslikavanja skupa $A$ u skup $B$ jednak je
$$|\{f| f: A\xrightarrow{\text{"na"}} B\}| = \sum\limits_{i=0}^{k-1}\dbinom{k}{i}(-1)^i(k-i)^n$$

_Dokaz:_

Neka je $B = \{b_1, b_2, ..., b_k\}. Preslikavanje $f: A\rightarrow B$$ koje nije sirjektivno ima osobinu da postoji element skupa $B$ koji nije u skupu slika elemenata iz skupa $A$ (u oznaci $im(f)$), tj. $f$ pripada jednom od sledećih skupova:
$$
\begin{aligned}
B_1 &= \{f: A\rightarrow B | b_1\notin im(f)\} \\
B_2 &= \{f: A\rightarrow B | b_2\notin im(f)\} \\
... \\
B_k &= \{f: A\rightarrow B | b_k\notin im(f)\} \\
\end{aligned}
$$

Skup svih preslikavanja skupa $A$ u skup $B$ možemo podeliti na skup svih preslikavanja koja su "na" i skup svih preslikavanja koja nisu "na". Tada je:
$$
\begin{aligned}
\{f| f: A\rightarrow B\} &= \{f | f: A\xrightarrow{\text{"na"}} B\} \cup \{f | f: A\xrightarrow{\text{nije "na"}} B\} \\
\{f| f: A\rightarrow B\} &= \{f | f: A\xrightarrow{\text{"na"}} B\} \cup B_1 \cup B_2 \cup ... \cup B_k
\end{aligned}
$$

Na osnovu principa sume sledi
$$|\{f| f: A\rightarrow B\}| = |\{f | f: A\xrightarrow{\text{"na"}} B\}| + |B_1 \cup B_2 \cup ... \cup B_k|$$

Primenom principa uključenja-isključenja:
$$|B_1 \cup B_2 \cup ... \cup B_k| = \sum\limits_{\emptyset \not= I\subseteq \{1,..., k\}}{(-1)^{|I|-1} \left| \bigcap\limits_{i\in I} B_i\right|}$$

Može se zaključiti da za sve $i, j, i_1, ..., i_k\in \{1,..., k\}$ važi:
$$
\begin{aligned}
|B_i| &= |\{f: A\rightarrow B\setminus\{b_i\}\}| = (k-1)^n \\
|B_i\cap B_j| &= |\{f: A\rightarrow B\setminus\{b_i, b_j\}\}| = (k-2)^n \\
&... \\ 
|B_{i_1} \cap...\cap B_{i_{k-1}}| &= |\{f: A\rightarrow B\setminus\{b_{i_1},...,b_{i_{k-1}}\}\}| = (k-(k-1))^n = 1 \\
|B_{i_1} \cap...\cap B_{i_{k}}| &= |\{f: A\rightarrow B\setminus\{b_{i_1},...,b_{i_{k}}\}\}| = (k-k)^n = 0 \\
\end{aligned}
$$
Tako dobijamo
$$|B_1 \cup B_2 \cup ... \cup B_k| = \dbinom{k}{1}(k-1)^n - \dbinom{k}{2}(k-2)^n + ... + (-1)^{k-2} $$

_Napomena:_ Binomni koeficijent se pojavljuje kao broj načina da odaberemo $i$ elemenata koji se neće nalaziti u skupu slika.

S obzirom da je ukupan broj preslikavanja skupa $A$ u skup $B$ jednak $k^n$, dobijamo da je:
$$
\begin{aligned}
|\{f| f: A\xrightarrow{\text{"na"}} B\}| &= |\{f| f: A\rightarrow B\}| - |B_1\cup B_2 \cup ... \cup B_k| \\
&= k^n - k(k-1)^n + \dbinom{k}{2}(n-2)^n - ... + (-1)^{k-1} \\
&= \sum\limits_{i = 0}^{n-1}{\dbinom{k}{i}(-1)^i(k-i)^n}
\end{aligned}
$$

<br/>

$$\tag*{□}$$

<br/>

```python
def sirjective(n, k):
    return sum([(-1)**i * bin_koef(k, i) * (k - i)**n for i in range(k + 1)])
```

```python
print(sirjective(3, 2))
```

## Stirlingovi brojevi II vrste

Stirlingovi brojevi II vrste (u oznaci $\left\{ n \atop k \right\}$ ili $S(n, k)$) predstavljaju broj mogućnosti da se skup od $n$ elemenata podeli na $k$ disjunktinih nepraznih skupova.

_Primer:_
$$
A = \{a, b, c, d\} \\
n = |A| = 4 \\
k = 3 \\
\left\{ 4 \atop 3 \right\} = 6
$$
Želimo da podelimo skup $A$ na 3 neprazna disjunktna skupa.
$$
\{a, b\}, \{c\}, \{d\} \\
\{a, c\}, \{b\}, \{d\} \\
\{a, d\}, \{b\}, \{c\} \\
\{a\}, \{b, c\}, \{d\} \\
\{a\}, \{b, d\}, \{c\} \\
\{a\}, \{b\}, \{c, d\} \\
$$

_Rekurzivna definicija_

$$S(n, k) = \underbrace{k\cdot S(n-1, k)}_{\text{1.}} + \underbrace{S(n-1, k-1)}_{\text{2.}}$$

1. Ako imamo $n$ elemenata, onda uzmemo jedan element i dodamo ga u jedan od $k$ podskupova, a ostalih $n-1$ elemenata rasporedimo u $k$ podskupova.
2. Kako smo uzeli jedan element i formurali jedan podskup, ostalo nam je $n-1$ elemanta rasporedimo u $k-1$ podskupova

_Bazni slučajevi:_
$S(n, 0) = 0, n\gt 0$ - ne možemo imati podskupove bez elemenata

$S(0, 0) = 1$ - jedan praz skupo može biti podeljen u 0 podskupova na 1 način

$S(n, n) = 1$ - svaki element je u svom podskupu


```python
def stirling_second_kind(n, k):
    if k == 0 or k > n:
        return 0
    if n == k or k == 1:
        return 1
    return k * stirling_second_kind(n - 1, k) + stirling_second_kind(n - 1, k - 1) 
```

```python
print(stirling_second_kind(4, 3)) # Output: 6
```

## Sirjektivne funkcije i Stirlingovi brojevi

Kada podelimo skup $A$ sa $n$ elemenata u $k$ nepraznih podskupa, svaki od njih se može posmatrati kao da odgovara po jednom elementu iz skupa $B$ sa $k$ elemenata. Svaki podskup možemo "mapirati" na jedan od elemenata skupa $B$ (što možemo uraditi na $k!$ načina). Ako je svaki podskup povezan sa rezličitim elementom iz skupa $B$, onda ćemo imati sirjektivnu funkciju.

_Primer:_

$$A=\{1,2,3\}\\
B = \{x, y\}$$

Podele skupa $A$ su:
$$
\begin{aligned}
P_1 &= \{\{1, 2\}, \{3\}\} \\
P_2 &= \{\{1, 3\}, \{2\}\} \\
P_3 &= \{\{1\}, \{2, 3\}\}
\end{aligned}
$$

Ove particije možemo "mapirati" na sledeće načine da bismo dobili sirjektivne funkcije:

Od $P_1$:
$$
\begin{aligned}
f_{11} &= \begin{pmatrix}
   1 & 2 & 3 \\
   x & x & y \\
\end{pmatrix} \\
f_{12} &= \begin{pmatrix}
   1 & 2 & 3 \\
   y & y & x \\
\end{pmatrix}
\end{aligned}
$$

Od $P_2$:
$$
\begin{aligned}
f_{21} &= \begin{pmatrix}
   1 & 2 & 3 \\
   x & y & x \\
\end{pmatrix} \\
f_{22} &= \begin{pmatrix}
   1 & 2 & 3 \\
   y & x & y \\
\end{pmatrix}
\end{aligned}
$$

Od $P_3$:
$$
\begin{aligned}
f_{31} &= \begin{pmatrix}
   1 & 2 & 3 \\
   x & y & y \\
\end{pmatrix} \\
f_{32} &= \begin{pmatrix}
   1 & 2 & 3 \\
   y & x & x \\
\end{pmatrix}
\end{aligned}
$$


Stoga je ukupan broj sirjektivnih funkcija skupa od $n$ elemenata na skup od $k$ elemenata jednak $$k!S(n, k)$$

Odnosno za date skupove $A = \{a_1, a_2, ..., a_n\} i $B = \{b_1, b_2, ..., b_k\}$ važi:

$$|\{f| f: A\xrightarrow{\text{"na"}} B\}| = k!\cdot S(n, k)$$


```python
import math

def sirjective_stirling(n, k):
    return math.factorial(k) * stirling_second_kind(n, k)
```

```python
A = {1, 2, 3}
B = {"x", "y"}
n = len(A)
k = len(B)
print(sirjective_stirling(n, k)) # Output: 6
```

## Algoritmi i primeri 
- Paskalov trougao, algoritmi za razvoj binomne i polinomne formule

### 1. Paskalov trougao

Bazira se na Paskalovom identitetu koji glasi:
    Za cele brojeve $n$ i $m$, $1 \leq m \leq n$, važi:


$ \binom{n}{k} = \binom{n-1}{m} + \binom{n-1}{m-1}$ 


```python
def pascal_triangle_half(n):
    triangle = []
    
    for i in range(n):
        row = [1]  
        if i > 0:
            for j in range(1, ((i // 2) + 1)):
                row.append(triangle[i-1][j-1] + triangle[i-1][j])
        triangle.append(row + row[-2::-1] if i % 2 == 0 else row + row[::-1])

    return triangle

def PrintT(triangle):
    max_width = len(" ".join(str(i) for i in triangle[-1]))  # calculate the width of the longest row
    for row in triangle:
        print(" ".join(str(i) for i in row).center(max_width))

n = 10
triangle = PrintT(pascal_triangle_half(n))

```

Prikaz Paskalovog trougla kao 'tabela'

```python
def pascal_triangle_table(n):
    triangle = []
    for i in range(n):
        row = [1]
        if i > 0:
            for j in range(1, i):
                row.append(triangle[i - 1][j - 1] + triangle[i - 1][j])
            row.append(1)
        triangle.append(row)
    
    max_num = triangle[-1][len(triangle[-1]) // 2]
    cell_width = len(str(max_num)) + 2  # Adjust +2 for padding

    max_cols = len(triangle[-1])

    header = [" " * cell_width] + [f"{i:^{cell_width}}" for i in range(max_cols)]
    print("".join(header))

    print(" " * cell_width + "_" * (cell_width * max_cols))

    for i, row in enumerate(triangle):
        row_str = [f"{i:^{cell_width}}"] + [f"{num:^{cell_width}}" for num in row]
        row_str += ["-" * cell_width] * (max_cols - len(row))  # Fill remaining cells with '-'
        print("|".join(row_str))

n = 10
pascal_triangle_table(n)
```

## Razvoj binomne formule

Podsetimo se matematičkog zapisa:

$ (x+y)^n = \displaystyle\sum_{m=0}^n \binom{n}{k} x ^ {n-m} y ^{m}$

```python
def factorial(x):
    if x == 0 or x == 1:
        return 1
    else:
        return x * factorial(x - 1)

def binomial_coefficient(n, k):
    return factorial(n) // (factorial(k) * factorial(n - k))

def binomial_expansion_latex(a, b, n):
    terms = []
    for k in range(n + 1):
        coeff = binomial_coefficient(n, k)
        term_a = f"{a}^{{{n-k}}}" if n - k > 0 else ""
        term_b = f"{b}^{{{k}}}" if k > 0 else ""
        
        if coeff != 1:
            term_str = f"{coeff} \\cdot {term_a}" if term_a else f"{coeff}"
            if term_b:
                term_str += f" \\cdot {term_b}"
        else:
            term_str = term_a if term_a else term_b
        
        terms.append(term_str)

    expansion = " + ".join(terms)
    return f"({a} + {b})^{n} = {expansion}"

a = 5
b = 9
n = 5
latex_expansion = binomial_expansion_latex(a, b, n)

from IPython.display import display, Markdown
display(Markdown(f"$$ {latex_expansion} $$"))
```

## Razvoj polinomne formule

Matematički zapis:

$ \binom{n}{m_1, m_2, ..., m_l} = \frac{n!}{m_1! \cdot ... \cdot m_l!} $

```python
def factorial(x):
    if x == 0 or x == 1:
        return 1
    else:
        return x * factorial(x - 1)

def multinomial_coefficient(n, k):
    coeff = factorial(n)
    for value in k:
        coeff //= factorial(value)
    return coeff

def polynomial_formula_latex(n, **kwargs):
    terms = []
    variables = list(kwargs.keys())
    m = len(variables)

    # Generate all possible combinations of exponents for the variables
    def generate_terms(exponents, total, index):
        if index == m:
            if sum(exponents) == n:
                coeff = multinomial_coefficient(n, exponents)
                term = ' \\cdot '.join(
                    f"{variables[i]}^{{{exponents[i]}}}" for i in range(m) if exponents[i] > 0
                )
                terms.append(f"{coeff} \\cdot {term}" if coeff != 1 else term)
            return

        for exp in range(n + 1):
            exponents[index] = exp
            generate_terms(exponents, total + exp, index + 1)

    generate_terms([0] * m, 0, 0)

    polynomial = " + ".join(terms)
    variables_string = " + ".join(variables)
    return f"({variables_string})^{n} = {polynomial}"

n = 2
polynomial_expansion_latex = polynomial_formula_latex(n, a=1, b=1, c=1, d=1)

from IPython.display import display, Markdown
display(Markdown(f"$$ {polynomial_expansion_latex} $$"))
```

### Vandermondov identitet

```python
def factorial(x):
    """Calculate factorial of x."""
    if x == 0 or x == 1:
        return 1
    else:
        return x * factorial(x - 1)

def binomial_coefficient(n, k):
    """Calculate the binomial coefficient C(n, k)."""
    if k < 0 or k > n:
        return 0
    return factorial(n) // (factorial(k) * factorial(n - k))

def vandermonde_identity(m, n, r):
    """Verify Vandermonde's identity: 
    sum_{k=0}^{r} C(m, k) * C(n, r-k) = C(m+n, r)"""
    
    lhs = sum(binomial_coefficient(m, k) * binomial_coefficient(n, r - k) for k in range(r + 1))
    
    rhs = binomial_coefficient(m + n, r)
    
    print(f"LHS: {lhs}, RHS: {rhs}")
    
    return lhs == rhs

# Example usage
m = 5
n = 3
r = 4
is_valid = vandermonde_identity(m, n, r)

if is_valid:
    print(f"Vandermonde's identity holds for m={m}, n={n}, r={r}.")
else:
    print(f"Vandermonde's identity does not hold for m={m}, n={n}, r={r}.")
```
