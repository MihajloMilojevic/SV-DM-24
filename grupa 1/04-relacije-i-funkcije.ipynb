# Relacije i funkcije

### Operacije sa funkcijama

S obzirom da su funkcije u stvari skupovi parova elemenata, sa njima možemo da vršimo sve operacije kao i sa skupovima. Međutim, pored njih, postoje i operacije koje su namenjene samo funkcijama.

#### Slaganje (kompozicija) funkcija

Ako su date funkcije $f: A\rightarrow B$ i $g: B\rightarrow C$, tada se pod kompozijom funkcija $f$ i $g$ podrazumeva funckija $g\circ f: A\rightarrow C$ data pravilom preslikavanja:
$$g\circ f = g(f(x)), x\in A$$

#### Redukcija funkcije

Ako su dati funckija $f: A\rightarrow B$ i podskup $A'\subseteq A$, tada se pod redukcijom funkcije $f$ na poddomen $A'$ u oznaci $f_{A'}$ podrazumeva funkcija $f_{A'}: A'\rightarrow B$, data pravilom preslikavanja:
$$f_{A'}(x) = f(x), x\in A'$$

#### Identička funkcija

Za proizvoljan skup $A$, funkcija $i_{A}: A\rightarrow A$ data pravilom preslikavanja $i_A(x) = x, x\in A$ naziva se identička funkcija na skupu $A$.

#### Inverzna funkcija

Ako je data funkcija $f: A\rightarrow B$, tada se za funkciju $g: B\rightarrow A$ kaže da je inverzna funkcija za funkciju $f$ ako važi:
$$g(f(x)) = x, x\in A$$
odnosno
$$f(g(y)) = y, y\in B$$

Inverzna funkcija za funkciju f se obično obeležava sa $f^{-1}$.


#### Teorema

Za funkciju $f: A\rightarrow B$ postoji inverztna funkcija ako i samo ako je $f$ bijekcija.

_Dokaz:_

$\implies$ Pretpostavimo da funkcija $f: A\rightarrow B$ ima inverznu funkciju $f^{-1}: B\rightarrow A$ i dokažimo da je $f$ bijekcija.

Funkcija f je injektivna, jer ako je $f(x_1)=f(x_2)=y$ za $x_1 \not = x_2$, tada iz definisanosti inverzne funkcije sledi da je $f^{-1}(y)=f^{-1}(x_1)=x_1$ i $f^{-1}(y)=f^{-1}(x_2)=x_2$, što je kontradiktorno činjenici da je $f^{-1}$ funkcija i da prema tome, može da ima samo jednu vrednost za datu vrednost argumenta. S druge strane funkcija je i sirjektivna, jer za svako $y\in B$ važi $f(f^{-1}(y)) = y$.

$\impliedby$ Pretpostavimo sada da je $f$ bijekcija i dokažimo da onda postoji funkcija $h$ koja je njena inverzna funkcija. Funkciju $h$ ćemo definisati na sledeći način:
$$h\subseteq B\times A, (y, x)\in h \text{ ako i samo ako je } f(x)=y$$

Primetimo da smo ovim, u stvari samo definisali jedan podskup $h\subseteq B\times A$, pa stoga moramo da pokažemo da je h zaista funkcija, tj. da za svako $y\in B$ postoji tačno jedno $x\in A$ tako da je $h(y) = x$. Najpre, pošto je funckija $f$ sirjektivna, to za svako $y\in B$ postoji $x\in A$ takvo da je $f(x) = y$, pa po definiciji $h$ važi i $(y, x)\in h$. Dalje, pošto je $f$ injektivna ovakvo $x$ je jedinstveno, pa smo se zaista uverili da je $h$ funkcija i možemo slobodno da pišemo $h: B\rightarrow A, h(y) = x$ ako i samo je $f(x) = y$. Sada je jasno da važi $h(f(x)) = x$ i $f(h(y)) = y$ pa je $h$ inverzna funckija za $f$.

## Relacije

### Osnovni pojmovi

Relacije predstavljaju veze (odnose) između izvesnih objekata. Relacija je neprazan podskup Dekartovog proizvoda skupova.

Relacija $\rho$ dužine $n$ je neprazan podskup Dekartovog proizvoda $n$ skupova. Kada je $n=2$ tada govorimo o binarnoj relaciji, dakle o relaciji između elementa $x$ sa elementom $y$, odnosno o uređenom paru $(x, y)$  iz Dekartovog proizvoda $A\times B$. Ako je $(x, y) \in \rho$ tada kažemo da je element $x$ u relaciji sa elementom $y$ i pišemo $x\rho y$.

Binarna relacija je samo sinonim za skup uređenih parova.

Za binarnu relaciju $\rho \subset A\times B$ moguće je definisati izraze:
- Domen - oblast definisanosti $D(\rho) = \{ x\in A | (\exists y\in B) x\rho y\}$
- Kodomen - oblast definisanosti $K(\rho) = \{y\in B | (\exists x \in A) x\rho y\}$
- Inverzna relacija $\rho^{-1} = \{(y, x) | \forall (x, y) \in \rho \}$
- Kompozicija relacija $\rho \circ \sigma$ ako je $\sigma \subset A\times B$ i $\rho \subset B\times C$. Tada je relacija $\rho \circ \sigma \subset A\times C$ data sa $\rho \circ \sigma = \{(x, y) | (\exists z \in B) x\sigma z \land z\rho y\}$ kompozicija relcija $\rho$ i $\sigma$.

### Primer 1

$$
A = \{1,3,6\} \\
A\times A = \{(1,1),(1,3),(1,6),(3,1),(3,3),(3,6),(6,1),(6,3),(6,6)\} \\
\rho = \{(1,1),(1,3),(1,6),(3,1),(3,3),(3,6),(6,6)\} \\
\rho \subseteq A^2
$$
U ovom slučaju relacija $\rho$ manje ili jednako $\le$ izgleda ovako: 
$$ (1\le 1),(1\le 3),(1\le 6),(3\le 1),(3\le 3),(3\le 6),(6\le 6) $$

### Osobine

1. Refleksivnost (R) - relacija je refleksivna ako i samo ako je svaki element u relaciji sa samim sobom
$$\rho \text{ je refleksivna } \iff (\forall x\in A) x\rho x$$

Primer bi bio da u grafu, u svakom čvoru tog grafa postoji petlja.

2. Simetričnost (S) - relacija je simetrična ako za svaki uređeni par elemenata koji je u relaciji postoji i par sa obrnutim poretkom
$$\rho \text{ je simetrična } \iff ((\forall x,y\in A): x\rho y \implies y\rho x)$$

Primer bi bio da je svaka grana u grafu dvosmerna.

3. Antisimetričnost (A) - relacija je antisimetrična ako ne postoje dva različita elementa koja su u relaciji sami sa sobom
$$\rho \text{ je antisimetrična } \iff ((\forall x, y\in A): (x\rho y \land y\rho x) \implies y = x)$$

4. Tranzitivnost (T) - relacija je tranzitivna ako važi da ako je prvi u relaciji sa drugim, drugi sa trećim onda mora biti i prvi u relaciji sa trećim
$$\rho \text{ je tranzitivna } \iff ((\forall x,y,z\in A): (x\rho y \land y\rho z) \implies x\rho z)$$

Primer bi bio da ako u usmerenom grafu postoji put između dva čvora onda postoji i duži put.

5. Funckija (F) - svaki element je u relaciji sa drugim elementom tačno jednom.
$$\rho \text{ je funkcija } \iff ((\forall x,y,z\in A): (x\rho y \land x\rho z) \implies y = z)$$


Posebne relacije su:
- relacija ekvivalencije koja je refleksivna, simetrična i tranzitivna (RST)
- relacija poretka koja je refleksivna, antisimetrična i tranzitivna (RAT)
- relacija strogog poretka: za bilo koji od $a, b \in A$ važi da $(a,b)\in \rho \lor (b, a) \in \rho$
- linearna ili totalna relacija: relacija poretka i relacija strogog poretka

### Primer 2

$$
A = \{1,2,3,4\} \\
\rho_1 = \{(1,1),(2,2),(3,3),(4,4)\} \\
\rho_2 = \{ (2,1),(1,1) \} \\
\rho_3 = \{(3,4),(4,2),(1,2) \} \\
\rho_4 = \{(1,1),(2,2),(3,3),(4,4),(3,4),(4,3) \} \\
\rho_5 = \{(1,1),(2,2),(3,3),(4,4),(3,4) \}
$$

Relacija $\rho_1$ je simetrična, antisimetrična, refleksivna, tranzitivna

Relacija $\rho_2$ je antisimetrična, tranzitivna

Relacija $\rho_3$ je antisimetrična

Relacija $\rho_4$ je refleksivna, simetrična, tranzitivna - relacija ekvivalencije

Relacija $\rho_5$ je refleksivna, antisimetrična, tranzitivna - relacije poretka


## Funkcije

### Matematička definicija

Pod funkcijom se podrazumeva uređena trojka $(A, B, f)$, gde je $f\subseteq A\times B$, pri čemu svako $x\in A$ postoji tačno jedno $y\in B$ tako da $(x, y)\in f$. Skup $A$ se naziva domen, skup $B$ se naziva kodomen, a $f$ je pravilo preslikavanja. Činjenica da funkcija preslikava elemente domena $A$ u elemente kodomena $B$ pomoću pravila preslikavanja $f$ zapisuje se pomoću $f:A\rightarrow B$, a za $(x, y) \in f$ ravnopravno i češće pišemo $f(x) = y$. Samo pravilo preslikavanja $f$ zadaje se formulom ili navođenjem parova elemenata $(x, y)$ koji pripadaju $f$.

### Osnovne osobine funkcija

#### Injekcija ("1-1")

Neka je $f: A\rightarrow B$. Funkcija $f$ je injektivna ("1-1") ako za svaka dva elementa $a\in A$ i $b\in B$ važi:
$$ a\not = b \implies f(a) \not = f(b) (\text{ ili čečće } f(a) = f(b) \implies a = b) $$

Injektivna funkcija iz $A$ u $B$ označava se sa:

$$f: A\xrightarrow{"1-1"} B$$

#### Sirjekcija ("na")

Neka je $f: A\rightarrow B$. Funkcija $f$ je sirjektivna ("na") ako je $f(A) = f(B)$, tj. za svaki element $c\in B$ postoji $a\in A$ sa osobinom $f(a) = c$:
$$(\forall c\in B)(\exists a\in A)(f(a)=c)$$

Sirjektivna funkcija iz $A$ u $B$ označava se sa:

$$f: A\xrightarrow{"na"} B$$

#### Bijekcija

Neka je $f: A\rightarrow B$. Funkcija $f$ je bijekcija (obostrano jednoznačno preslikavanje) ako je funkcija i injektivna i sirjektivna.

Bijektivna funkcija iz $A$ u $B$ označava se sa:

$$f: A\xrightarrow["na"]{"1-1"} B$$

#### Primer

Funkcija $a: \{1,2\}\rightarrow \{1,3,5\}$ data pravilom preslikavanja $a(x) = 2x - 1$ jeste injektivna, ali nije sirjektivna, jer se ni jedan element iz domena ne slika u element $5$ iz koodomena.

Funkcija $b: \{1,2,3\}\rightarrow \{1,3\}$ data pravilom preslikavanja $b = \{(1,1), (2,3), (3,1)\}$ jeste sirjektivna ali nije injektivna jer se elementi $1$ i $3$ oba slikau u isti element koodomena, $1$.

Funkcija $c: \{1,2,3\}\rightarrow \{1,3,5\}$ data pravilom preslikavanja $c(x) = 2x-1$ jeste i injektivna i sirjektivna, odnosno funkcija $c$ je bijekcija.

### Slikovno predstavljanje funkcija

Funkcija $f: A\rightarrow B$ se može slikovno predstaviti tako što se najpre predstave domen $A$ i kodomen $B$ funkcije, a onda se usmerenim linijama svaki element $x\in A$ poveže sa vrednošću $f(x)\in B$.
